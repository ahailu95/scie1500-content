## 1. The Antiderivative Concept

### What is an Antiderivative?

An **antiderivative** of $f(x)$ is a function $F(x)$ such that:

$$F'(x) = f(x)$$

In other words, the antiderivative is the function whose derivative gives us back the original function.

### The Question of Uniqueness

If $F(x)$ is an antiderivative of $f(x)$, then so is $F(x) + C$ for any constant $C$.

**Why?** Because the derivative of a constant is zero:
$$\frac{d}{dx}[F(x) + C] = F'(x) + 0 = f(x)$$

This leads to the **constant of integration** $C$.

```python
# Visualizing multiple antiderivatives

# f(x) = 2x has antiderivatives F(x) = x² + C
x = np.linspace(-3, 3, 200)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Left: The function f(x) = 2x
ax1.plot(x, 2*x, 'b-', linewidth=2, label='f(x) = 2x')
ax1.axhline(y=0, color='gray', linewidth=0.5)
ax1.axvline(x=0, color='gray', linewidth=0.5)
ax1.set_xlabel('x', fontsize=12)
ax1.set_ylabel('f(x)', fontsize=12)
ax1.set_title('The function f(x) = 2x', fontsize=14)
ax1.legend()
ax1.grid(True, alpha=0.3)

# Right: Family of antiderivatives F(x) = x² + C
colors = ['red', 'orange', 'green', 'blue', 'purple']
C_values = [-4, -2, 0, 2, 4]

for C, color in zip(C_values, colors):
    ax2.plot(x, x**2 + C, color=color, linewidth=2, label=f'F(x) = x² + {C}' if C >= 0 else f'F(x) = x² - {-C}')

ax2.axhline(y=0, color='gray', linewidth=0.5)
ax2.axvline(x=0, color='gray', linewidth=0.5)
ax2.set_xlabel('x', fontsize=12)
ax2.set_ylabel('F(x)', fontsize=12)
ax2.set_title('Antiderivatives: F(x) = x² + C', fontsize=14)
ax2.legend(loc='upper center', fontsize=9)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("Key Insight: All these curves have the SAME slope at any given x!")
print("The constant C shifts the curve vertically without changing its slope.")
```

## 2. Indefinite Integrals

### Notation

The **indefinite integral** of $f(x)$ is written as:

$$\int f(x) \, dx = F(x) + C$$

where:
- $\int$ is the integral sign
- $f(x)$ is the **integrand**
- $dx$ indicates integration with respect to $x$
- $F(x)$ is an antiderivative of $f(x)$
- $C$ is the **constant of integration**

### Connection to Differentiation

Integration and differentiation are inverse operations:

$$\frac{d}{dx}\left[\int f(x) \, dx\right] = f(x)$$

$$\int f'(x) \, dx = f(x) + C$$

## 3. The Power Rule for Integration

### The Rule

For $n \neq -1$:

$$\int x^n \, dx = \frac{x^{n+1}}{n+1} + C$$

**Memory aid:** "Add 1 to the power, divide by the new power"

### Important Examples

| Integrand | Integral |
|-----------|----------|
| $x^3$ | $\frac{x^4}{4} + C$ |
| $x^2$ | $\frac{x^3}{3} + C$ |
| $x$ | $\frac{x^2}{2} + C$ |
| $1 = x^0$ | $x + C$ |
| $x^{-2} = \frac{1}{x^2}$ | $\frac{x^{-1}}{-1} = -\frac{1}{x} + C$ |
| $\sqrt{x} = x^{1/2}$ | $\frac{x^{3/2}}{3/2} = \frac{2}{3}x^{3/2} + C$ |

```python
# Examples of the power rule

print("Power Rule Examples: ∫ xⁿ dx = x^(n+1)/(n+1) + C")
print("="*55)
print()

examples = [
    (5, "x⁵", "x⁶/6"),
    (2, "x²", "x³/3"),
    (1, "x", "x²/2"),
    (0, "1", "x"),
    (-2, "x⁻² = 1/x²", "-1/x"),
    (0.5, "x^(1/2) = √x", "(2/3)x^(3/2)")
]

for n, integrand, result in examples:
    if n != -1:
        print(f"n = {n:>4}:  ∫ {integrand:>12} dx = {result:>15} + C")

print()
print("Note: The power rule does NOT work for n = -1.")
print("      ∫ x⁻¹ dx = ∫ 1/x dx = ln|x| + C (special case!)")
```

## 4. Basic Integration Rules

### Constant Multiple Rule

$$\int k \cdot f(x) \, dx = k \int f(x) \, dx$$

### Sum/Difference Rule

$$\int [f(x) \pm g(x)] \, dx = \int f(x) \, dx \pm \int g(x) \, dx$$

```python
# Example: Integrating a polynomial

print("Example: Find ∫ (6x² - 4x + 5) dx")
print("="*45)
print()
print("Apply the sum rule to integrate term by term:")
print()
print("∫ 6x² dx = 6 · (x³/3) = 2x³")
print("∫ (-4x) dx = -4 · (x²/2) = -2x²")
print("∫ 5 dx = 5x")
print()
print("Therefore:")
print("∫ (6x² - 4x + 5) dx = 2x³ - 2x² + 5x + C")
print()

# Verify by differentiation
print("Verification (differentiate the answer):")
print("d/dx (2x³ - 2x² + 5x + C) = 6x² - 4x + 5 ✓")
```

## 5. Integrals of Exponential Functions

### The Natural Exponential

$$\int e^x \, dx = e^x + C$$

### Exponential with Linear Argument

$$\int e^{kx} \, dx = \frac{1}{k} e^{kx} + C$$

### General Exponential

$$\int a^x \, dx = \frac{a^x}{\ln a} + C \quad (a > 0, a \neq 1)$$

```python
# Visualizing exponential integration

x = np.linspace(-2, 2, 200)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Left: e^x and its integral
ax1.plot(x, np.exp(x), 'b-', linewidth=2, label='f(x) = eˣ')
ax1.plot(x, np.exp(x), 'r--', linewidth=2, label='∫eˣ dx = eˣ + C (same curve!)')
ax1.set_xlabel('x', fontsize=12)
ax1.set_ylabel('y', fontsize=12)
ax1.set_title('eˣ is its own integral (up to constant)', fontsize=14)
ax1.legend()
ax1.grid(True, alpha=0.3)

# Right: e^(2x) and its integral
ax2.plot(x, np.exp(2*x), 'b-', linewidth=2, label='f(x) = e^(2x)')
ax2.plot(x, 0.5*np.exp(2*x), 'r-', linewidth=2, label='∫e^(2x) dx = (1/2)e^(2x) + C')
ax2.set_xlabel('x', fontsize=12)
ax2.set_ylabel('y', fontsize=12)
ax2.set_title('Integrating e^(kx) divides by k', fontsize=14)
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("Examples:")
print("  ∫ e^(3x) dx = (1/3)e^(3x) + C")
print("  ∫ e^(-x) dx = -e^(-x) + C")
print("  ∫ e^(x/2) dx = 2e^(x/2) + C")
```

## 6. The Special Case: Integral of 1/x

### Why the Power Rule Fails

For $n = -1$, the power rule would give:
$$\int x^{-1} \, dx = \frac{x^0}{0} = \frac{1}{0} \quad \text{(undefined!)}$$

### The Natural Logarithm

Instead, we have:

$$\int \frac{1}{x} \, dx = \ln|x| + C$$

The absolute value ensures the result is defined for negative $x$ as well.

```python
# Visualizing ∫ 1/x dx = ln|x|

x_pos = np.linspace(0.1, 5, 200)
x_neg = np.linspace(-5, -0.1, 200)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Left: f(x) = 1/x
ax1.plot(x_pos, 1/x_pos, 'b-', linewidth=2, label='f(x) = 1/x')
ax1.plot(x_neg, 1/x_neg, 'b-', linewidth=2)
ax1.axhline(y=0, color='gray', linewidth=0.5)
ax1.axvline(x=0, color='gray', linewidth=0.5)
ax1.set_xlabel('x', fontsize=12)
ax1.set_ylabel('f(x)', fontsize=12)
ax1.set_title('f(x) = 1/x', fontsize=14)
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.set_ylim(-5, 5)

# Right: F(x) = ln|x|
ax2.plot(x_pos, np.log(x_pos), 'r-', linewidth=2, label='∫(1/x)dx = ln|x| + C')
ax2.plot(x_neg, np.log(np.abs(x_neg)), 'r-', linewidth=2)
ax2.axhline(y=0, color='gray', linewidth=0.5)
ax2.axvline(x=0, color='gray', linewidth=0.5)
ax2.set_xlabel('x', fontsize=12)
ax2.set_ylabel('F(x)', fontsize=12)
ax2.set_title('F(x) = ln|x| (the antiderivative)', fontsize=14)
ax2.legend()
ax2.grid(True, alpha=0.3)
ax2.set_ylim(-3, 3)

plt.tight_layout()
plt.show()

print("Important: ln|x| is defined for both positive and negative x.")
print("  For x > 0: ln|x| = ln(x)")
print("  For x < 0: ln|x| = ln(-x)")
```

## 7. Finding Particular Solutions with Initial Conditions

### The Problem

The indefinite integral gives a **family** of solutions (all differing by a constant $C$).

To find a **particular** solution, we need an **initial condition** — a known point $(x_0, y_0)$ on the curve.

### The Process

1. Find the general antiderivative: $F(x) + C$
2. Use the initial condition to solve for $C$
3. Write the particular solution

```python
# Example: Finding a particular solution

print("PROBLEM: Find f(x) given that f'(x) = 3x² - 4x and f(2) = 5")
print("="*60)
print()
print("Step 1: Find the general antiderivative")
print("  f(x) = ∫ (3x² - 4x) dx")
print("       = x³ - 2x² + C")
print()
print("Step 2: Use the initial condition f(2) = 5")
print("  5 = (2)³ - 2(2)² + C")
print("  5 = 8 - 8 + C")
print("  5 = C")
print()
print("Step 3: Write the particular solution")
print("  f(x) = x³ - 2x² + 5")
print()

# Visualization
x = np.linspace(-1, 4, 200)

def f_general(x, C):
    return x**3 - 2*x**2 + C

plt.figure(figsize=(10, 6))

# Plot several members of the family
for C in [-5, 0, 5, 10]:
    alpha = 1.0 if C == 5 else 0.3
    lw = 2 if C == 5 else 1
    label = f'f(x) = x³ - 2x² + {C}' if C == 5 else None
    color = 'blue' if C == 5 else 'gray'
    plt.plot(x, f_general(x, C), color=color, linewidth=lw, alpha=alpha, label=label)

# Mark the initial condition
plt.scatter([2], [5], color='red', s=100, zorder=5, label='Initial condition: (2, 5)')

plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.title('Finding the Particular Solution Through (2, 5)', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## 8. Application: From Velocity to Position

### The Physics Connection

If $v(t)$ is velocity and $s(t)$ is position, then:

$$s(t) = \int v(t) \, dt$$

This is because velocity is the derivative of position: $v(t) = s'(t)$.

```python
# Example: Finding position from velocity

print("PROBLEM: A particle has velocity v(t) = 6t - 2 m/s.")
print("         If the initial position is s(0) = 10 m, find s(t).")
print("="*60)
print()
print("Step 1: Integrate velocity to get position")
print("  s(t) = ∫ (6t - 2) dt")
print("       = 3t² - 2t + C")
print()
print("Step 2: Use initial condition s(0) = 10")
print("  10 = 3(0)² - 2(0) + C")
print("  C = 10")
print()
print("Step 3: Particular solution")
print("  s(t) = 3t² - 2t + 10")
print()

# Visualization
t = np.linspace(0, 4, 200)
v = 6*t - 2
s = 3*t**2 - 2*t + 10

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

ax1.plot(t, v, 'b-', linewidth=2, label='v(t) = 6t - 2')
ax1.axhline(y=0, color='gray', linewidth=0.5)
ax1.fill_between(t, v, where=(v > 0), alpha=0.2, color='green', label='Positive velocity')
ax1.fill_between(t, v, where=(v < 0), alpha=0.2, color='red', label='Negative velocity')
ax1.set_xlabel('Time t (s)', fontsize=12)
ax1.set_ylabel('Velocity v(t) (m/s)', fontsize=12)
ax1.set_title('Velocity Function', fontsize=14)
ax1.legend()
ax1.grid(True, alpha=0.3)

ax2.plot(t, s, 'r-', linewidth=2, label='s(t) = 3t² - 2t + 10')
ax2.scatter([0], [10], color='blue', s=100, zorder=5, label='s(0) = 10')
ax2.set_xlabel('Time t (s)', fontsize=12)
ax2.set_ylabel('Position s(t) (m)', fontsize=12)
ax2.set_title('Position Function (Integral of Velocity)', fontsize=14)
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("At t = 2s:")
print(f"  Velocity: v(2) = 6(2) - 2 = 10 m/s")
print(f"  Position: s(2) = 3(4) - 2(2) + 10 = 18 m")
```

## Key Formulas Summary

|-------|-------------|

| Constant multiple | $\int k \cdot f(x) \, dx = k \int f(x) \, dx$ |
| Sum/difference | $\int [f \pm g] \, dx = \int f \, dx \pm \int g \, dx$ |
