## Part A: Function Basics (~20 min)

### What is a function?

A **function** $f$ assigns to each valid input $x$ exactly one output $f(x)$.  
We write: $f : \text{Domain} \to \text{Range}$

- The **domain** is the set of all valid inputs.
- The **range** is the set of all possible outputs.

For the plastic production data, we define:

$$P(t) = \text{production (million tonnes)}, \quad t = \text{years since 1950}$$

So $P(0) = 2$, $P(20) = 35$, $P(65) = 380$, etc.


```python
# The data from the UNEP report
t_values = np.array([0, 20, 40, 60, 65])   # years since 1950
P_values = np.array([2, 35, 120, 270, 380]) # million tonnes

# Quick check — what does t = 60 correspond to?
years = 1950 + t_values
print("Year:       ", years)
print("t (yrs):    ", t_values)
print("P(t) (Mt):  ", P_values)
```

### A.1 Average Rate of Change

The **average rate of change** of $P$ from $t = a$ to $t = b$ is:

$$\frac{P(b) - P(a)}{b - a}$$

This is the slope of the straight line joining the two points.  
Its units here are: **million tonnes per year**.


```python
# Average rate of change from 1950 to 2015 (t = 0 to t = 65)
P_1950 = 2    # P(0)
P_2015 = 380  # P(65)
delta_t = 65

avg_rate = (P_2015 - P_1950) / delta_t
print(f"Average rate of change (1950–2015): {avg_rate:.2f} million tonnes per year")
```

## Part B: Interpreting Slope (~15 min)

The slope tells us *how fast* production is growing.  
Does it grow at the same rate in every decade, or does the rate change?


```python
# Calculate average rate of change between consecutive data points
decades = [(0, 20, '1950–1970'), (20, 40, '1970–1990'),
           (40, 60, '1990–2010'), (60, 65, '2010–2015')]

print(f"{'Period':<12}  {'Rate (Mt/yr)':>14}")
print("-" * 30)
for t1, t2, label in decades:
    P1 = P_values[t_values == t1][0]
    P2 = P_values[t_values == t2][0]
    rate = (P2 - P1) / (t2 - t1)
    print(f"{label:<12}  {rate:>14.2f}")
```

### B.2 Plotting P(t)

Let's see the shape of $P(t)$ directly.


```python
fig, ax = plt.subplots()
ax.scatter(1950 + t_values, P_values, color='steelblue', s=80, zorder=5, label='Data points')
ax.plot(1950 + t_values, P_values, color='steelblue', linewidth=1.5, alpha=0.6)

ax.set_xlabel('Year')
ax.set_ylabel('Production (million tonnes)')
ax.set_title('Global Plastic Production, 1950–2015')
ax.legend()
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

## Part C: Australian Policy Application (~20 min)

Now we apply what we know about functions and rates of change to a real policy question.

**Context:**  
Australia's coastal population in 2010 was approximately 20 million, with each person generating  
50 kg of plastic waste per year. Of this waste, **2% is mismanaged** (enters the ocean).

**Define $t$ = years since 2010 for this part.**


```python
# Given values
population_2010 = 20e6        # 20 million people
plastic_per_person_kg = 50    # kg/person/year
mismanagement_rate = 0.02     # 2%

# Step 1: Annual mismanaged plastic waste entering the ocean (in 2010)
annual_waste_kg = population_2010 * plastic_per_person_kg * mismanagement_rate
annual_waste_tonnes = annual_waste_kg / 1000   # convert kg → tonnes

print(f"Annual mismanaged plastic (2010): {annual_waste_tonnes:,.0f} tonnes/year")
print(f"Daily mismanaged plastic (2010):  {annual_waste_tonnes/365:,.1f} tonnes/day")
```

### C.1 Writing the Function W(t)

If Australia's coastal population grows at **1.5% per year**, then at time $t$ years after 2010:

$$\text{Population}(t) = 20{,}000{,}000 \times 1.015^t$$

✏️ **Activity C.1:** Write the full function $W(t)$ for annual mismanaged plastic waste (in tonnes/year)  
as a function of $t$, then implement it in Python.


```python
def W(t):
    """Annual mismanaged plastic waste (tonnes/year), t years after 2010."""
    growth_rate = 1.015          # 1.5% annual population growth
    # YOUR CODE: complete the formula
    population_t = 20e6 * growth_rate**t
    waste = population_t * plastic_per_person_kg * mismanagement_rate / 1000
    return waste

# Test: W(0) should match your 2010 value above
print(f"W(0)  = {W(0):,.0f} tonnes/year  (should match 2010 value above)")
print(f"W(10) = {W(10):,.0f} tonnes/year  (2020 projection)")
print(f"W(20) = {W(20):,.0f} tonnes/year  (2030 projection)")
```

### C.2 Policy Scenario

The government proposes reducing the mismanagement rate from **2% to 0.5%** by 2030.  
Would this reduction offset the effect of population growth?

✏️ **Activity C.2:** Define a new function $W_{\text{policy}}(t)$ with the lower mismanagement rate  
and compare it to $W(t)$ at $t = 20$ (year 2030).


```python
def W_policy(t):
    """Annual mismanaged waste (tonnes/year) under 0.5% mismanagement policy."""
    growth_rate = 1.015
    mismanagement_new = 0.005    # 0.5%
    population_t = 20e6 * growth_rate**t
    waste = population_t * plastic_per_person_kg * mismanagement_new / 1000
    return waste

# Compare at 2030 (t = 20)
t_2030 = 20
print(f"W(20)  — business as usual:    {W(t_2030):,.0f} tonnes/year")
print(f"W_policy(20) — policy applied: {W_policy(t_2030):,.0f} tonnes/year")
print()
# Does the policy offset population growth?
# Compare W_policy(20) to W(0) — the 2010 baseline
print(f"2010 baseline W(0):             {W(0):,.0f} tonnes/year")
diff = W_policy(t_2030) - W(0)
direction = "INCREASE" if diff > 0 else "DECREASE"
print(f"\nUnder the policy, 2030 waste would {direction} by {abs(diff):,.0f} tonnes/year vs 2010.")
```

### C.3 Visualise the Two Scenarios


```python
t_range = np.linspace(0, 30, 100)

fig, ax = plt.subplots()
ax.plot(2010 + t_range, [W(t) for t in t_range],
        label='Business as usual (2% mismanagement)', color='tomato', linewidth=2)
ax.plot(2010 + t_range, [W_policy(t) for t in t_range],
        label='Policy scenario (0.5% mismanagement)', color='steelblue', linewidth=2, linestyle='--')

ax.axvline(2010, color='gray', linestyle=':', alpha=0.7)
ax.set_xlabel('Year')
ax.set_ylabel('Mismanaged plastic (tonnes/year)')
ax.set_title("Australia's Ocean Plastic Waste — Two Scenarios")
ax.legend()
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
