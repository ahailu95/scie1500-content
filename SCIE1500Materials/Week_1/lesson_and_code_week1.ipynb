## 2. What Is a Function?

A **function** $f$ is a rule that assigns to each input value $x$ (from a set called the **domain**) exactly one output value $f(x)$ (in a set called the **range**).

$$f: \text{Domain} \to \text{Range}$$

### The Vertical Line Test

A graph represents a function if and only if **no vertical line intersects the curve more than once**.

### Domain and Range

- **Domain $D$**: The set of all valid input values
- **Range $R$**: The set of all possible output values

**Interval Notation:**
- Closed: $[a, b]$ means $a \leq x \leq b$
- Open: $(a, b)$ means $a < x < b$

### Examples

| Function | Domain | Range |
|----------|--------|-------|
| $f(x) = x^2$ | $\mathbb{R}$ | $y \geq 0$ |
| $f(x) = \sqrt{x}$ | $x \geq 0$ | $y \geq 0$ |
| $f(x) = 2^x$ | $\mathbb{R}$ | $y > 0$ |

```python
# The Vertical Line Test: Function vs Not-a-Function

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Left: Parabola y = x^2 (IS a function)
x = np.linspace(-3, 3, 200)
y = x**2
ax1.plot(x, y, 'b-', linewidth=2.5, label='$y = x^2$')
# Vertical line test
ax1.axvline(x=1.5, color='red', linestyle='--', linewidth=1.5, alpha=0.7, label='Vertical line')
ax1.plot(1.5, 1.5**2, 'ro', markersize=10, zorder=5)
ax1.set_xlabel('x', fontsize=12)
ax1.set_ylabel('y', fontsize=12)
ax1.set_title('Parabola: PASSES vertical line test ✓', fontsize=13)
ax1.legend(fontsize=11)
ax1.grid(True, alpha=0.3)
ax1.set_xlim(-3, 3)
ax1.set_ylim(-1, 9)

# Right: Circle x^2 + y^2 = 4 (NOT a function)
theta = np.linspace(0, 2*np.pi, 200)
cx = 2 * np.cos(theta)
cy = 2 * np.sin(theta)
ax2.plot(cx, cy, 'b-', linewidth=2.5, label='$x^2 + y^2 = 4$')
# Vertical line hits circle twice
ax2.axvline(x=1.0, color='red', linestyle='--', linewidth=1.5, alpha=0.7, label='Vertical line')
y_intersect = np.sqrt(4 - 1.0**2)
ax2.plot([1.0, 1.0], [y_intersect, -y_intersect], 'ro', markersize=10, zorder=5)
ax2.set_xlabel('x', fontsize=12)
ax2.set_ylabel('y', fontsize=12)
ax2.set_title('Circle: FAILS vertical line test ✗', fontsize=13)
ax2.legend(fontsize=11)
ax2.grid(True, alpha=0.3)
ax2.set_xlim(-3, 3)
ax2.set_ylim(-3, 3)
ax2.set_aspect('equal')

plt.tight_layout()
plt.show()

print("A parabola is a function: each x gives exactly one y.")
print("A circle is NOT a function: some x values give TWO y values.")
```

## 3. Linear Functions

### General Form

$$f(x) = mx + c$$

where $m$ is the **slope** (rate of change) and $c$ is the **y-intercept**.

- **Domain:** $\mathbb{R}$ (all real numbers)
- **Range:** $\mathbb{R}$

### Scientific Meaning

Linear functions describe **constant rates of change**. If a quantity increases by the same amount in each time period, the relationship is linear.

**Example:** Suppose we model early plastic production as:
$$P(t) = 2 + 0.5t \quad \text{(million tonnes, } t = 0 \text{ is 1950)}$$

At $t = 0$: $P(0) = 2$ million tonnes. At $t = 20$: $P(20) = 12$ million tonnes.
But actual 1970 production was 35 million tonnes — linear growth underestimates reality.

```python
# Defining and Plotting Linear Functions

def linear_fun(x, m, c):
    """Linear function: f(x) = mx + c"""
    return m * x + c

# Create a range of x values
x = np.linspace(-5, 5, 200)

# Plot several linear functions with different slopes
slopes = [(-2, 'Negative slope (m = -2)'), 
          (0, 'Zero slope (m = 0)'), 
          (0.5, 'Small positive (m = 0.5)'), 
          (2, 'Steep positive (m = 2)')]

plt.figure(figsize=(10, 6))
colors = ['red', 'gray', 'blue', 'green']

for (m, label), color in zip(slopes, colors):
    y = linear_fun(x, m, c=1)
    plt.plot(x, y, color=color, linewidth=2, label=label)

plt.axhline(y=0, color='black', linewidth=0.5)
plt.axvline(x=0, color='black', linewidth=0.5)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x) = mx + 1', fontsize=12)
plt.title('Linear Functions: How Slope Affects the Graph', fontsize=14)
plt.legend(loc='best', fontsize=10)
plt.grid(True, alpha=0.3)
plt.xlim(-5, 5)
plt.ylim(-8, 12)
plt.tight_layout()
plt.show()

# Demonstrate evaluation
print("Evaluating f(x) = 2x + 1:")
for x_val in [0, 1, 2, 3]:
    print(f"  f({x_val}) = 2({x_val}) + 1 = {linear_fun(x_val, 2, 1)}")
```

## 4. Quadratic Functions

### General Form

$$f(x) = ax^2 + bx + c$$

### Key Features

1. **Vertex (turning point):** $x = -\frac{b}{2a}$
2. **Concavity:** Opens upward if $a > 0$, downward if $a < 0$
3. **X-intercepts (roots):** $x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$
4. **Y-intercept:** $c$ (when $x = 0$)

### Sketching a Quadratic

1. Find x-intercepts: set $y = 0$ and solve
2. Find y-intercept: set $x = 0$
3. Find vertex: $x_{vertex} = -\frac{b}{2a}$, compute $y_{vertex}$

### Example: $y = x^2 - 2x - 8$

**X-intercepts:** $(x-4)(x+2) = 0 \Rightarrow x = 4, x = -2$
**Y-intercept:** $y = -8$
**Vertex:** $x = 1$, $y = 1 - 2 - 8 = -9$ → Vertex at $(1, -9)$

```python
# Quadratic Functions: Vertex, Intercepts, and Concavity

def quad_fun(x, a, b, c):
    """Quadratic function: f(x) = ax^2 + bx + c"""
    return a * x**2 + b * x + c

# Example: y = x^2 - 2x - 8
a, b, c = 1, -2, -8
x = np.linspace(-4, 6, 300)
y = quad_fun(x, a, b, c)

# Calculate key features
x_vertex = -b / (2 * a)
y_vertex = quad_fun(x_vertex, a, b, c)
discriminant = b**2 - 4*a*c
x_root1 = (-b + np.sqrt(discriminant)) / (2*a)
x_root2 = (-b - np.sqrt(discriminant)) / (2*a)

plt.figure(figsize=(10, 7))
plt.plot(x, y, 'b-', linewidth=2.5, label=f'$y = x^2 - 2x - 8$')

# Mark key features
plt.scatter([x_vertex], [y_vertex], color='red', s=150, zorder=5, label=f'Vertex ({x_vertex}, {y_vertex})')
plt.scatter([x_root1, x_root2], [0, 0], color='green', s=100, zorder=5, label=f'Roots: x = {x_root2:.0f}, {x_root1:.0f}')
plt.scatter([0], [c], color='orange', s=100, zorder=5, label=f'Y-intercept: (0, {c})')

# Annotations
plt.annotate(f'Vertex\n({x_vertex:.0f}, {y_vertex:.0f})', (x_vertex, y_vertex),
             xytext=(x_vertex + 1.5, y_vertex + 1), fontsize=10,
             arrowprops=dict(arrowstyle='->', color='red'))

plt.axhline(y=0, color='black', linewidth=0.5)
plt.axvline(x=0, color='black', linewidth=0.5)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('Quadratic Function: Key Features', fontsize=14)
plt.legend(loc='upper right', fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("Key Features of y = x² - 2x - 8:")
print(f"  Vertex: ({x_vertex:.0f}, {y_vertex:.0f})")
print(f"  X-intercepts: x = {x_root2:.0f} and x = {x_root1:.0f}")
print(f"  Y-intercept: y = {c}")
print(f"  Concavity: Opens {'upward' if a > 0 else 'downward'} (a = {a})")
```

## 5. Function Symmetry

### Even Functions

A function is **even** if $f(-x) = f(x)$ for all $x$ in the domain.
- Graph is symmetric about the **y-axis**
- Example: $f(x) = x^2$

### Odd Functions

A function is **odd** if $f(-x) = -f(x)$ for all $x$ in the domain.
- Graph is symmetric about the **origin**
- Example: $f(x) = x^3$

Understanding symmetry helps identify function types from graphical data and simplifies analysis.

```python
# Visualising Even and Odd Functions

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

x = np.linspace(-3, 3, 300)

# Even function: f(x) = x^2
y_even = x**2
ax1.plot(x, y_even, 'b-', linewidth=2.5, label='$f(x) = x^2$')
# Show symmetry: f(-2) = f(2) = 4
ax1.plot([-2, 2], [4, 4], 'ro', markersize=10, zorder=5)
ax1.plot([-2, 2], [4, 4], 'r--', linewidth=1.5, alpha=0.5)
ax1.annotate('f(-2) = 4', (-2, 4), xytext=(-2.8, 5.5), fontsize=10,
             arrowprops=dict(arrowstyle='->', color='red'))
ax1.annotate('f(2) = 4', (2, 4), xytext=(2.2, 5.5), fontsize=10,
             arrowprops=dict(arrowstyle='->', color='red'))
ax1.axvline(x=0, color='gray', linewidth=1, linestyle='-', alpha=0.3)
ax1.set_xlabel('x', fontsize=12)
ax1.set_ylabel('y', fontsize=12)
ax1.set_title('Even Function: f(-x) = f(x)', fontsize=13)
ax1.legend(fontsize=11)
ax1.grid(True, alpha=0.3)

# Odd function: f(x) = x^3
y_odd = x**3
ax2.plot(x, y_odd, 'g-', linewidth=2.5, label='$f(x) = x^3$')
# Show symmetry: f(-2) = -f(2)
ax2.plot([-2], [-8], 'ro', markersize=10, zorder=5)
ax2.plot([2], [8], 'ro', markersize=10, zorder=5)
ax2.annotate('f(-2) = -8', (-2, -8), xytext=(-2.8, -15), fontsize=10,
             arrowprops=dict(arrowstyle='->', color='red'))
ax2.annotate('f(2) = 8', (2, 8), xytext=(2.2, 15), fontsize=10,
             arrowprops=dict(arrowstyle='->', color='red'))
ax2.axhline(y=0, color='gray', linewidth=0.5)
ax2.axvline(x=0, color='gray', linewidth=0.5)
ax2.set_xlabel('x', fontsize=12)
ax2.set_ylabel('y', fontsize=12)
ax2.set_title('Odd Function: f(-x) = -f(x)', fontsize=13)
ax2.legend(fontsize=11)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Test symmetry programmatically
print("Testing symmetry:")
print(f"\nEven test for f(x) = x²:")
for val in [1, 2, 3]:
    print(f"  f({val}) = {val**2},  f({-val}) = {(-val)**2}  →  Equal? {val**2 == (-val)**2}")

print(f"\nOdd test for f(x) = x³:")
for val in [1, 2, 3]:
    print(f"  f({val}) = {val**3},  f({-val}) = {(-val)**3}  →  f(-x) = -f(x)? {(-val)**3 == -(val**3)}")
```

## 6. Domain Restrictions in Scientific Modelling

**Getting domains right is essential for valid science.** A model predicting negative concentrations or undefined values is scientifically meaningless.

### Common Mathematical Restrictions

| Expression Type | Restriction | Example |
|-----------------|-------------|---------|
| $\frac{1}{g(x)}$ | $g(x) \neq 0$ | $\frac{1}{x-2}$: $x \neq 2$ |
| $\sqrt{g(x)}$ | $g(x) \geq 0$ | $\sqrt{x-3}$: $x \geq 3$ |
| $\ln(g(x))$ | $g(x) > 0$ | $\ln(x+1)$: $x > -1$ |

### Physical Domain Constraints

Beyond mathematical restrictions, physical constraints often limit domains:
- Time: $t \geq 0$
- Population: $P \geq 0$
- Concentration: $C \geq 0$
- Proportions: $0 \leq p \leq 1$

```python
# Visualising Domain Restrictions

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# 1/x: undefined at x = 0
x1 = np.linspace(-5, -0.1, 200)
x2 = np.linspace(0.1, 5, 200)
axes[0].plot(x1, 1/x1, 'b-', linewidth=2)
axes[0].plot(x2, 1/x2, 'b-', linewidth=2)
axes[0].axvline(x=0, color='red', linestyle='--', linewidth=2, label='x ≠ 0')
axes[0].set_title('$f(x) = 1/x$\nDomain: $x \\neq 0$', fontsize=12)
axes[0].set_ylim(-10, 10)
axes[0].legend(fontsize=10)
axes[0].grid(True, alpha=0.3)

# sqrt(x): defined for x >= 0
x_pos = np.linspace(0, 10, 200)
axes[1].plot(x_pos, np.sqrt(x_pos), 'g-', linewidth=2.5)
axes[1].axvspan(-2, 0, alpha=0.2, color='red', label='Undefined region')
axes[1].scatter([0], [0], color='green', s=100, zorder=5)
axes[1].set_title('$f(x) = \\sqrt{x}$\nDomain: $x \\geq 0$', fontsize=12)
axes[1].set_xlim(-2, 10)
axes[1].legend(fontsize=10)
axes[1].grid(True, alpha=0.3)

# ln(x): defined for x > 0
x_ln = np.linspace(0.01, 10, 200)
axes[2].plot(x_ln, np.log(x_ln), 'purple', linewidth=2.5)
axes[2].axvspan(-2, 0, alpha=0.2, color='red', label='Undefined region')
axes[2].axvline(x=0, color='red', linestyle='--', linewidth=1.5)
axes[2].set_title('$f(x) = \\ln(x)$\nDomain: $x > 0$', fontsize=12)
axes[2].set_xlim(-2, 10)
axes[2].set_ylim(-4, 3)
axes[2].legend(fontsize=10)
axes[2].grid(True, alpha=0.3)

for ax in axes:
    ax.set_xlabel('x', fontsize=11)
    ax.set_ylabel('f(x)', fontsize=11)

plt.tight_layout()
plt.show()

print("Domain Restriction Summary:")
print("  f(x) = 1/x      → Domain: x ≠ 0 (division by zero)")
print("  f(x) = √x       → Domain: x ≥ 0 (square root of negative)")
print("  f(x) = ln(x)    → Domain: x > 0 (log of non-positive)")
```

## 7. Real-World Application: Ocean Plastic Sampling

### The Reisser Correction Model

Reisser et al. (2013) developed a correction function for ocean plastic sampling. Surface trawls capture plastic at the surface, but plastic distributes throughout the water column. The ratio of plastic concentration at depth $d$ to surface concentration is:

$$\frac{C_i}{C_s} = \frac{1}{1 - e^{-d \cdot \alpha}}$$

where $\alpha = w_b / A_0$ combines buoyant velocity and turbulence. This function has domain restriction $d > 0$ (depth must be positive).

```python
# The Reisser Correction Model for Ocean Plastic Sampling

def correction_ratio(d, alpha=0.4):
    """
    Ratio of depth-integrated to surface concentration.
    d: sampling depth (metres)
    alpha: buoyancy/turbulence parameter
    """
    return 1 / (1 - np.exp(-d * alpha))

# Sampling depths
d = np.linspace(0.01, 10, 500)

# Plot the correction function
plt.figure(figsize=(10, 6))
plt.plot(d, correction_ratio(d, 0.2), 'b-', linewidth=2, label='α = 0.2 (low turbulence)')
plt.plot(d, correction_ratio(d, 0.4), 'r-', linewidth=2, label='α = 0.4 (moderate)')
plt.plot(d, correction_ratio(d, 0.8), 'g-', linewidth=2, label='α = 0.8 (high turbulence)')
plt.axhline(y=1, color='gray', linestyle='--', alpha=0.5, label='Ratio = 1 (deep water)')

plt.xlabel('Sampling Depth d (metres)', fontsize=12)
plt.ylabel('Correction Ratio $C_i / C_s$', fontsize=12)
plt.title('Reisser Correction: How Depth Affects Concentration Estimate', fontsize=14)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.ylim(0.5, 8)
plt.tight_layout()
plt.show()

# Calculate specific values
print("Correction ratios at different depths (α = 0.4):")
for depth in [0.5, 1.0, 2.0, 5.0, 10.0]:
    ratio = correction_ratio(depth, 0.4)
    print(f"  Depth {depth:>4.1f}m: Multiply surface count by {ratio:.3f}")
print("\nKey insight: Shallow sampling overestimates concentration (ratio > 1).")
print("As depth increases, the correction approaches 1.")
```

## 8. Visualising Global Plastic Production

As analytical scientists, we don't just calculate — we **visualise**. Let's examine the global plastics production data to see why we need different function types.

```python
# Global Plastic Production Data (1950-2015)
# Source: Geyer et al. (2017), "Production, use, and fate of all plastics ever made"

years = np.array([1950, 1955, 1960, 1965, 1970, 1975, 1980, 1985, 
                  1990, 1995, 2000, 2005, 2010, 2015])
production = np.array([2, 5, 8, 15, 35, 50, 70, 90, 
                       120, 150, 200, 250, 300, 380])  # million metric tonnes

# Create time variable (t = 0 in 1950)
t = years - 1950

# Plot the data
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Left: Raw data
ax1.plot(years, production, 'ro-', markersize=8, linewidth=2, label='Actual data')

# Overlay a linear fit for comparison
m_linear = (production[-1] - production[0]) / (t[-1] - t[0])
linear_fit = production[0] + m_linear * t
ax1.plot(years, linear_fit, 'b--', linewidth=1.5, label=f'Linear: P ≈ {production[0]} + {m_linear:.1f}t')

ax1.set_xlabel('Year', fontsize=12)
ax1.set_ylabel('Production (Million Metric Tonnes)', fontsize=12)
ax1.set_title('Global Plastic Production (1950–2015)', fontsize=13)
ax1.legend(fontsize=10)
ax1.grid(True, alpha=0.3)

# Right: Log scale to check for exponential growth
ax2.semilogy(years, production, 'go-', markersize=8, linewidth=2, label='Data (log scale)')
ax2.set_xlabel('Year', fontsize=12)
ax2.set_ylabel('Production (log scale)', fontsize=12)
ax2.set_title('Log Scale: Is Growth Exponential?', fontsize=13)
ax2.legend(fontsize=10)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("Key observations:")
print("  1. Growth is clearly NOT linear — production accelerates over time")
print("  2. On a log scale, the data is roughly linear → suggests exponential growth")
print("  3. This motivates our study of exponential functions in Week 2!")
```
