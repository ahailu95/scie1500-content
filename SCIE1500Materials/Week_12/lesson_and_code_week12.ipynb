## 1. Systems of Linear Equations

### Two Equations, Two Unknowns

A system of two linear equations:

$$a_1x + b_1y = c_1$$
$$a_2x + b_2y = c_2$$

### Solution Methods

1. **Substitution:** Solve one equation for one variable, substitute
2. **Elimination:** Add/subtract equations to eliminate a variable
3. **Graphical:** Find intersection of the two lines

```python
# Example: Solve the system
# 2x + 3y = 12
# x - y = 1

print("Solving the System:")
print("  2x + 3y = 12")
print("  x - y = 1")
print("="*40)
print()

# Method 1: Substitution
print("Method 1: Substitution")
print("-"*30)
print("From equation 2: x = y + 1")
print("Substitute into equation 1:")
print("  2(y + 1) + 3y = 12")
print("  2y + 2 + 3y = 12")
print("  5y = 10")
print("  y = 2")
print("Then x = y + 1 = 3")
print()

# Verify using numpy
A = np.array([[2, 3], [1, -1]])
b = np.array([12, 1])
solution = np.linalg.solve(A, b)

print(f"Solution: (x, y) = ({solution[0]:.1f}, {solution[1]:.1f})")
print()

# Graphical verification
x = np.linspace(-1, 6, 100)

# Rearrange to y = ...
y1 = (12 - 2*x) / 3  # 2x + 3y = 12
y2 = x - 1            # x - y = 1

plt.figure(figsize=(10, 6))
plt.plot(x, y1, 'b-', linewidth=2, label='2x + 3y = 12')
plt.plot(x, y2, 'r-', linewidth=2, label='x - y = 1')
plt.scatter([3], [2], color='green', s=150, zorder=5, label=f'Solution (3, 2)')

plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('System of Linear Equations - Graphical Solution', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='gray', linewidth=0.5)
plt.axvline(x=0, color='gray', linewidth=0.5)
plt.xlim(-1, 6)
plt.ylim(-2, 5)
plt.show()
```

## 2. Market Equilibrium

### Supply and Demand

- **Demand function:** $Q_d = f(P)$ — typically decreasing (higher price → lower demand)
- **Supply function:** $Q_s = g(P)$ — typically increasing (higher price → higher supply)

### Equilibrium

**Market equilibrium** occurs where supply equals demand:

$$Q_s = Q_d$$

The solution gives the **equilibrium price** $P^*$ and **equilibrium quantity** $Q^*$.

```python
# Market equilibrium example

# Demand: P = 100 - 2Q  (or Q_d = 50 - 0.5P)
# Supply: P = 20 + Q    (or Q_s = P - 20)

print("Market Equilibrium Problem")
print("="*45)
print()
print("Demand: P = 100 - 2Q")
print("Supply: P = 20 + Q")
print()
print("At equilibrium: Demand Price = Supply Price")
print("  100 - 2Q = 20 + Q")
print("  80 = 3Q")
print("  Q* = 80/3 ≈ 26.67")
print()
print("Equilibrium price:")
print("  P* = 20 + Q* = 20 + 26.67 ≈ $46.67")
print()

Q_eq = 80/3
P_eq = 20 + Q_eq

print(f"Equilibrium: (Q*, P*) = ({Q_eq:.2f}, ${P_eq:.2f})")

# Visualize
Q = np.linspace(0, 50, 200)
P_demand = 100 - 2*Q
P_supply = 20 + Q

plt.figure(figsize=(10, 6))
plt.plot(Q, P_demand, 'b-', linewidth=2, label='Demand: P = 100 - 2Q')
plt.plot(Q, P_supply, 'r-', linewidth=2, label='Supply: P = 20 + Q')
plt.scatter([Q_eq], [P_eq], color='green', s=150, zorder=5, label=f'Equilibrium ({Q_eq:.1f}, ${P_eq:.1f})')

plt.xlabel('Quantity (Q)', fontsize=12)
plt.ylabel('Price (P)', fontsize=12)
plt.title('Market Equilibrium', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.xlim(0, 50)
plt.ylim(0, 110)
plt.show()
```

## 3. Consumer and Producer Surplus

### Consumer Surplus

The benefit consumers receive from paying less than their maximum willingness to pay:

$$CS = \int_0^{Q^*} D(Q) \, dQ - P^* \cdot Q^*$$

Geometrically: Area between demand curve and equilibrium price line.

### Producer Surplus

The benefit producers receive from selling above their minimum acceptable price:

$$PS = P^* \cdot Q^* - \int_0^{Q^*} S(Q) \, dQ$$

Geometrically: Area between equilibrium price line and supply curve.

```python
# Calculate Consumer and Producer Surplus

print("Consumer and Producer Surplus")
print("="*50)
print()

# From previous: P_demand = 100 - 2Q, P_supply = 20 + Q
# Equilibrium: Q* = 80/3, P* = 140/3

Q_star = 80/3
P_star = 140/3

# Consumer Surplus: Triangle with vertices at (0, 100), (0, P*), (Q*, P*)
# Height = 100 - P* = 100 - 140/3 = 160/3
# Base = Q* = 80/3
# Area = 0.5 * base * height
CS = 0.5 * Q_star * (100 - P_star)

print("Consumer Surplus:")
print(f"  CS = ½ × Q* × (P_max - P*)")
print(f"     = ½ × {Q_star:.2f} × ({100} - {P_star:.2f})")
print(f"     = ${CS:.2f}")
print()

# Producer Surplus: Triangle with vertices at (0, 20), (0, P*), (Q*, P*)
# Height = P* - 20 = 140/3 - 20 = 80/3
# Base = Q* = 80/3
PS = 0.5 * Q_star * (P_star - 20)

print("Producer Surplus:")
print(f"  PS = ½ × Q* × (P* - P_min)")
print(f"     = ½ × {Q_star:.2f} × ({P_star:.2f} - {20})")
print(f"     = ${PS:.2f}")
print()

print(f"Total Surplus = CS + PS = ${CS + PS:.2f}")

# Visualization
Q = np.linspace(0, 50, 200)
P_demand = 100 - 2*Q
P_supply = 20 + Q

Q_fill = np.linspace(0, Q_star, 100)
P_demand_fill = 100 - 2*Q_fill
P_supply_fill = 20 + Q_fill

plt.figure(figsize=(12, 7))

# Plot curves
plt.plot(Q, P_demand, 'b-', linewidth=2, label='Demand')
plt.plot(Q, P_supply, 'r-', linewidth=2, label='Supply')

# Shade CS (between demand and P*)
plt.fill_between(Q_fill, P_star, P_demand_fill, alpha=0.3, color='blue', label=f'CS = ${CS:.0f}')

# Shade PS (between P* and supply)
plt.fill_between(Q_fill, P_supply_fill, P_star, alpha=0.3, color='red', label=f'PS = ${PS:.0f}')

# Equilibrium point
plt.scatter([Q_star], [P_star], color='green', s=150, zorder=5)
plt.axhline(y=P_star, color='green', linestyle='--', alpha=0.5)
plt.axvline(x=Q_star, color='green', linestyle='--', alpha=0.5)

plt.annotate(f'Equilibrium\n({Q_star:.1f}, ${P_star:.1f})', 
             xy=(Q_star, P_star), xytext=(Q_star+5, P_star+10),
             arrowprops=dict(arrowstyle='->', color='green'),
             fontsize=11)

plt.xlabel('Quantity (Q)', fontsize=12)
plt.ylabel('Price (P)', fontsize=12)
plt.title('Consumer and Producer Surplus', fontsize=14)
plt.legend(loc='upper right')
plt.grid(True, alpha=0.3)
plt.xlim(0, 50)
plt.ylim(0, 110)
plt.show()
```

## 4. Introduction to Linear Programming

### What is Linear Programming?

**Linear Programming (LP)** is a method to find the best outcome (maximum or minimum) in a mathematical model with:
- A **linear objective function** to optimize
- **Linear inequality constraints**

### Standard Form

**Maximize (or Minimize):** $Z = c_1x + c_2y$

**Subject to:**
- $a_{11}x + a_{12}y \leq b_1$
- $a_{21}x + a_{22}y \leq b_2$
- $x \geq 0, \, y \geq 0$

---

## 5. Graphical Method for LP

### Steps

1. **Graph each constraint** as a line (replace ≤ with =)
2. **Shade the feasible region** satisfying all constraints
3. **Identify corner points** (vertices) of the feasible region
4. **Evaluate the objective function** at each corner point
5. **The optimal solution** is at the corner point giving the best value

```python
# Example LP Problem

print("Linear Programming Example")
print("="*50)
print()
print("A factory produces two products: A and B")
print("  - Product A: $3 profit, uses 2 hours labor, 1 kg material")
print("  - Product B: $5 profit, uses 1 hour labor, 2 kg material")
print("  - Available: 40 hours labor, 30 kg material")
print()
print("Maximize profit.")
print()

print("Mathematical Formulation:")
print("-"*40)
print("Let x = units of A, y = units of B")
print()
print("Maximize: Z = 3x + 5y")
print()
print("Subject to:")
print("  2x + y ≤ 40  (labor)")
print("  x + 2y ≤ 30  (material)")
print("  x ≥ 0, y ≥ 0")
```

```python
# Graphical solution

# Constraint lines (as y = ...)
x = np.linspace(0, 35, 300)

# 2x + y ≤ 40  →  y ≤ 40 - 2x
y_labor = 40 - 2*x

# x + 2y ≤ 30  →  y ≤ (30 - x)/2
y_material = (30 - x) / 2

# Feasible region
y_feasible = np.minimum(y_labor, y_material)
y_feasible = np.maximum(y_feasible, 0)

plt.figure(figsize=(12, 8))

# Plot constraints
plt.plot(x, y_labor, 'b-', linewidth=2, label='2x + y = 40 (Labor)')
plt.plot(x, y_material, 'r-', linewidth=2, label='x + 2y = 30 (Material)')

# Feasible region
plt.fill_between(x, 0, y_feasible, where=(y_feasible > 0) & (x >= 0), 
                 alpha=0.3, color='green', label='Feasible Region')

# Corner points
corners = [
    (0, 0),      # Origin
    (20, 0),     # x-intercept of labor constraint
    (0, 15),     # y-intercept of material constraint
]

# Find intersection of constraints
# 2x + y = 40 and x + 2y = 30
# From first: y = 40 - 2x
# Substitute: x + 2(40 - 2x) = 30 → x + 80 - 4x = 30 → -3x = -50 → x = 50/3
# y = 40 - 2(50/3) = 40 - 100/3 = 20/3
x_int = 50/3
y_int = 20/3
corners.append((x_int, y_int))

# Evaluate objective at corners
print("Corner Point Analysis")
print("="*45)
print("Corner Point       Z = 3x + 5y")
print("-"*35)

best_z = -np.inf
best_corner = None

for xi, yi in corners:
    z = 3*xi + 5*yi
    print(f"({xi:5.2f}, {yi:5.2f})      Z = {z:.2f}")
    if z > best_z:
        best_z = z
        best_corner = (xi, yi)
    plt.scatter([xi], [yi], color='purple', s=100, zorder=5)

print("-"*35)
print(f"Optimal: ({best_corner[0]:.2f}, {best_corner[1]:.2f}) with Z = {best_z:.2f}")

# Highlight optimal point
plt.scatter([best_corner[0]], [best_corner[1]], color='gold', s=300, marker='*', zorder=6, 
            label=f'Optimal: ({best_corner[0]:.1f}, {best_corner[1]:.1f})')

# Iso-profit lines
for z_val in [30, 60, best_z]:
    y_iso = (z_val - 3*x) / 5
    style = '-' if z_val == best_z else '--'
    alpha = 1 if z_val == best_z else 0.5
    plt.plot(x, y_iso, 'g' + style, linewidth=1.5, alpha=alpha)

plt.xlabel('Units of Product A (x)', fontsize=12)
plt.ylabel('Units of Product B (y)', fontsize=12)
plt.title('Linear Programming: Graphical Solution', fontsize=14)
plt.legend(loc='upper right')
plt.grid(True, alpha=0.3)
plt.xlim(-1, 35)
plt.ylim(-1, 25)
plt.show()
```

```python
# Interpretation of the solution

print("Solution Interpretation")
print("="*50)
print()
print(f"Optimal production plan:")
print(f"  Product A: {best_corner[0]:.2f} ≈ 17 units")
print(f"  Product B: {best_corner[1]:.2f} ≈ 7 units")
print()
print(f"Maximum profit: Z = ${best_z:.2f}")
print()

# Check resource usage
labor_used = 2*best_corner[0] + best_corner[1]
material_used = best_corner[0] + 2*best_corner[1]

print("Resource utilization:")
print(f"  Labor: {labor_used:.2f} / 40 hours ({labor_used/40*100:.0f}%)")
print(f"  Material: {material_used:.2f} / 30 kg ({material_used/30*100:.0f}%)")
print()
print("Both constraints are binding (tight) at the optimal solution.")
```

## 6. Special Cases in LP

### Unbounded Solution
The feasible region extends infinitely in the direction of optimization.

### No Feasible Solution
The constraints are contradictory — no point satisfies all constraints.

### Multiple Optimal Solutions
When the objective function is parallel to a binding constraint, an infinite number of optimal solutions exist along that edge.

```python
# Example: Multiple optimal solutions

print("Example: Multiple Optimal Solutions")
print("="*50)
print()
print("Maximize: Z = 2x + 4y")
print("Subject to:")
print("  x + 2y ≤ 10")
print("  x + y ≤ 7")
print("  x, y ≥ 0")
print()

# Notice: The objective function 2x + 4y = Z
# gives y = (Z - 2x)/4 = Z/4 - x/2
# Slope = -1/2

# The constraint x + 2y = 10 gives y = 5 - x/2
# This has the same slope! So the objective is parallel to a constraint.

x = np.linspace(0, 12, 200)

# Constraints
y1 = (10 - x) / 2  # x + 2y = 10
y2 = 7 - x         # x + y = 7

# Feasible region
y_feas = np.minimum(y1, y2)
y_feas = np.maximum(y_feas, 0)

plt.figure(figsize=(10, 8))

plt.plot(x, y1, 'b-', linewidth=2, label='x + 2y = 10')
plt.plot(x, y2, 'r-', linewidth=2, label='x + y = 7')
plt.fill_between(x, 0, y_feas, where=(y_feas > 0) & (x >= 0), 
                 alpha=0.3, color='green', label='Feasible Region')

# Corner points
corners = [(0, 0), (7, 0), (0, 5), (4, 3)]

print("Corner Point Analysis:")
for xi, yi in corners:
    z = 2*xi + 4*yi
    print(f"  ({xi}, {yi}): Z = {z}")
    plt.scatter([xi], [yi], color='purple', s=100, zorder=5)

# Optimal edge
x_edge = np.linspace(0, 4, 50)
y_edge = (10 - x_edge) / 2
plt.plot(x_edge, y_edge, 'gold', linewidth=4, label='Optimal edge (Z = 20)')

print()
print("Multiple optimal solutions exist on the edge from (0, 5) to (4, 3).")
print("All points on this edge give Z = 20.")

plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('Multiple Optimal Solutions', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.xlim(-0.5, 10)
plt.ylim(-0.5, 8)
plt.show()
```

## 7. Sensitivity Analysis

### Shadow Prices

The **shadow price** of a constraint is the change in the optimal objective value per unit increase in the constraint's right-hand side.

- If a constraint is binding: shadow price > 0
- If a constraint has slack: shadow price = 0

```python
# Using scipy.optimize.linprog for LP

print("Solving LP with scipy.optimize.linprog")
print("="*50)
print()

# Original problem:
# Maximize: Z = 3x + 5y
# Subject to:
#   2x + y ≤ 40
#   x + 2y ≤ 30
#   x, y ≥ 0

# linprog minimizes, so we negate the objective
c = [-3, -5]  # Coefficients for minimization

# Inequality constraints: A_ub @ x <= b_ub
A_ub = [[2, 1],   # Labor constraint
        [1, 2]]   # Material constraint
b_ub = [40, 30]

# Bounds: x >= 0, y >= 0
x_bounds = (0, None)
y_bounds = (0, None)

result = linprog(c, A_ub=A_ub, b_ub=b_ub, bounds=[x_bounds, y_bounds], method='highs')

print(f"Optimal solution:")
print(f"  x = {result.x[0]:.4f}")
print(f"  y = {result.x[1]:.4f}")
print(f"  Maximum Z = {-result.fun:.4f}")
print()
print("(Matches our graphical solution!)")
```

| Market equilibrium | $Q_s = Q_d$ |
| Consumer surplus | $CS = \frac{1}{2} \times Q^* \times (P_{max} - P^*)$ (for linear demand) |

| LP standard form | Max/Min $Z = c_1x + c_2y$ subject to linear constraints |
