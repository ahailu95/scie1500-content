## 1. Derivatives of Exponential Functions

### The Natural Exponential Function

The function $f(x) = e^x$ is its own derivative:

$$\frac{d}{dx}(e^x) = e^x$$

This remarkable property makes $e^x$ fundamentally important in mathematics.

### General Exponential Functions

For $f(x) = a^x$ where $a > 0$:

$$\frac{d}{dx}(a^x) = a^x \ln(a)$$

```python
# Visualizing the derivative of e^x

x = np.linspace(-2, 2, 200)
y = np.exp(x)
dy = np.exp(x)  # derivative is the same!

fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(x, y, 'b-', linewidth=2, label='f(x) = eˣ')
ax.plot(x, dy, 'r--', linewidth=2, label="f'(x) = eˣ")

# Show tangent lines at a few points
for x_pt in [-1, 0, 1]:
    y_pt = np.exp(x_pt)
    slope = np.exp(x_pt)  # slope = function value!
    tangent_x = np.linspace(x_pt - 0.5, x_pt + 0.5, 50)
    tangent_y = y_pt + slope * (tangent_x - x_pt)
    ax.plot(tangent_x, tangent_y, 'g-', alpha=0.5, linewidth=1.5)
    ax.scatter([x_pt], [y_pt], color='green', s=60, zorder=5)

ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_title('The Unique Property: eˣ is its own derivative', fontsize=14)
ax.legend()
ax.grid(True, alpha=0.3)
ax.set_ylim(0, 8)
plt.show()

print("Key Insight: At every point on y = eˣ, the slope equals the y-value!")
```

## 2. Derivatives of Logarithmic Functions

### Natural Logarithm

$$\frac{d}{dx}(\ln x) = \frac{1}{x}$$

### General Logarithm

$$\frac{d}{dx}(\log_a x) = \frac{1}{x \ln a}$$

```python
# Visualizing the derivative of ln(x)

x = np.linspace(0.1, 5, 200)
y = np.log(x)      # ln(x)
dy = 1/x           # derivative

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Left: ln(x)
ax1.plot(x, y, 'b-', linewidth=2, label='f(x) = ln(x)')
ax1.axhline(y=0, color='gray', linewidth=0.5)
ax1.axvline(x=1, color='gray', linewidth=0.5, linestyle=':')
ax1.scatter([1], [0], color='blue', s=100, zorder=5)
ax1.set_xlabel('x', fontsize=12)
ax1.set_ylabel('y', fontsize=12)
ax1.set_title('f(x) = ln(x)', fontsize=14)
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.set_xlim(0, 5)
ax1.set_ylim(-2, 2)

# Right: derivative 1/x
ax2.plot(x, dy, 'r-', linewidth=2, label="f'(x) = 1/x")
ax2.axhline(y=0, color='gray', linewidth=0.5)
ax2.scatter([1], [1], color='red', s=100, zorder=5)
ax2.set_xlabel('x', fontsize=12)
ax2.set_ylabel('y', fontsize=12)
ax2.set_title("f'(x) = 1/x", fontsize=14)
ax2.legend()
ax2.grid(True, alpha=0.3)
ax2.set_xlim(0, 5)
ax2.set_ylim(0, 5)

plt.tight_layout()
plt.show()

print("Note: At x = 1, ln(1) = 0 and the slope of ln(x) is 1/1 = 1")
```

## 3. The Product Rule

### When Two Functions are Multiplied

If $f(x) = u(x) \cdot v(x)$, then:

$$f'(x) = u'(x) \cdot v(x) + u(x) \cdot v'(x)$$

**Memory aid:** "First derivative times second, plus first times second derivative"

```python
# Example: Product Rule

print("Example: Find the derivative of f(x) = x² · eˣ")
print("="*50)
print()
print("Let u(x) = x² and v(x) = eˣ")
print("Then u'(x) = 2x and v'(x) = eˣ")
print()
print("Product Rule: f'(x) = u'v + uv'")
print("             = 2x · eˣ + x² · eˣ")
print("             = eˣ(2x + x²)")
print("             = eˣ · x(2 + x)")
print()

# Numerical verification
def f(x):
    return x**2 * np.exp(x)

def f_prime(x):
    return np.exp(x) * (2*x + x**2)

x_test = 1
h = 0.0001
numerical = (f(x_test + h) - f(x_test)) / h

print(f"Verification at x = {x_test}:")
print(f"  f'({x_test}) = e¹ · (2·1 + 1²) = e · 3 = {f_prime(x_test):.4f}")
print(f"  Numerical: {numerical:.4f}")
```

## 4. The Quotient Rule

### When One Function Divides Another

If $f(x) = \frac{u(x)}{v(x)}$, then:

$$f'(x) = \frac{u'(x) \cdot v(x) - u(x) \cdot v'(x)}{[v(x)]^2}$$

**Memory aid:** "Low d-high minus high d-low, over low squared"

```python
# Example: Quotient Rule

print("Example: Find the derivative of f(x) = (x² + 1)/(x - 1)")
print("="*55)
print()
print("Let u(x) = x² + 1 (high) and v(x) = x - 1 (low)")
print("Then u'(x) = 2x and v'(x) = 1")
print()
print("Quotient Rule: f'(x) = [u'v - uv'] / v²")
print("             = [(2x)(x-1) - (x²+1)(1)] / (x-1)²")
print("             = [2x² - 2x - x² - 1] / (x-1)²")
print("             = [x² - 2x - 1] / (x-1)²")
print()

# Plot the function and its derivative
x_left = np.linspace(-2, 0.9, 100)
x_right = np.linspace(1.1, 4, 100)

def f(x):
    return (x**2 + 1) / (x - 1)

def f_prime(x):
    return (x**2 - 2*x - 1) / (x - 1)**2

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

ax1.plot(x_left, f(x_left), 'b-', linewidth=2)
ax1.plot(x_right, f(x_right), 'b-', linewidth=2, label='f(x) = (x² + 1)/(x - 1)')
ax1.axvline(x=1, color='red', linestyle='--', alpha=0.5, label='Vertical asymptote')
ax1.set_xlabel('x')
ax1.set_ylabel('f(x)')
ax1.set_title('f(x) = (x² + 1)/(x - 1)', fontsize=12)
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.set_ylim(-20, 20)

ax2.plot(x_left, f_prime(x_left), 'r-', linewidth=2)
ax2.plot(x_right, f_prime(x_right), 'r-', linewidth=2, label="f'(x)")
ax2.axvline(x=1, color='gray', linestyle='--', alpha=0.5)
ax2.axhline(y=0, color='gray', linewidth=0.5)
ax2.set_xlabel('x')
ax2.set_ylabel("f'(x)")
ax2.set_title("f'(x) = (x² - 2x - 1)/(x - 1)²", fontsize=12)
ax2.legend()
ax2.grid(True, alpha=0.3)
ax2.set_ylim(-5, 5)

plt.tight_layout()
plt.show()
```

## 5. The Chain Rule

### For Composite Functions

If $f(x) = g(h(x))$ (function inside a function), then:

$$f'(x) = g'(h(x)) \cdot h'(x)$$

**Memory aid:** "Derivative of the outer (evaluated at the inner) times derivative of the inner"

### Important Applications

| Function | Derivative |
|----------|------------|
| $e^{u(x)}$ | $e^{u(x)} \cdot u'(x)$ |
| $\ln(u(x))$ | $\frac{u'(x)}{u(x)}$ |
| $[u(x)]^n$ | $n[u(x)]^{n-1} \cdot u'(x)$ |

```python
# Example: Chain Rule

print("Example 1: Find the derivative of f(x) = e^(3x²)")
print("="*50)
print()
print("Let u(x) = 3x² (the inner function)")
print("Then u'(x) = 6x")
print("Outer function is eᵘ, with derivative eᵘ")
print()
print("Chain Rule: f'(x) = e^(3x²) · 6x = 6x · e^(3x²)")
print()

print("\nExample 2: Find the derivative of f(x) = ln(x² + 1)")
print("="*50)
print()
print("Let u(x) = x² + 1 (the inner function)")
print("Then u'(x) = 2x")
print("Outer function is ln(u), with derivative 1/u")
print()
print("Chain Rule: f'(x) = [1/(x² + 1)] · 2x = 2x/(x² + 1)")
print()

print("\nExample 3: Find the derivative of f(x) = (2x + 5)⁴")
print("="*50)
print()
print("Let u(x) = 2x + 5 (the inner function)")
print("Then u'(x) = 2")
print("Outer function is u⁴, with derivative 4u³")
print()
print("Chain Rule: f'(x) = 4(2x + 5)³ · 2 = 8(2x + 5)³")
```

## 6. Critical Points and Extrema

### Finding Where Functions Have Maximum or Minimum Values

**Critical points** occur where $f'(x) = 0$ or $f'(x)$ is undefined.

### First Derivative Test

At a critical point $x = c$:
- If $f'(x)$ changes from **positive to negative**: **Local maximum**
- If $f'(x)$ changes from **negative to positive**: **Local minimum**
- If $f'(x)$ doesn't change sign: **Neither** (inflection point)

### Second Derivative Test

At a critical point $x = c$ where $f'(c) = 0$:
- If $f''(c) < 0$: **Local maximum** (concave down)
- If $f''(c) > 0$: **Local minimum** (concave up)
- If $f''(c) = 0$: **Test is inconclusive** (use first derivative test)

```python
# Finding and classifying critical points

def f(x):
    return x**3 - 6*x**2 + 9*x + 2

def f_prime(x):
    return 3*x**2 - 12*x + 9

def f_double_prime(x):
    return 6*x - 12

# Find critical points: f'(x) = 3x² - 12x + 9 = 0
# 3(x² - 4x + 3) = 0
# 3(x - 1)(x - 3) = 0
# x = 1 or x = 3

critical_points = [1, 3]

print("Finding critical points for f(x) = x³ - 6x² + 9x + 2")
print("="*55)
print()
print("f'(x) = 3x² - 12x + 9 = 3(x² - 4x + 3) = 3(x - 1)(x - 3)")
print("f'(x) = 0 when x = 1 or x = 3")
print()

print("Using the Second Derivative Test:")
print("f''(x) = 6x - 12")
print()

for c in critical_points:
    second_deriv = f_double_prime(c)
    if second_deriv < 0:
        classification = "LOCAL MAXIMUM"
    elif second_deriv > 0:
        classification = "LOCAL MINIMUM"
    else:
        classification = "INCONCLUSIVE"
    print(f"At x = {c}: f''({c}) = 6({c}) - 12 = {second_deriv}")
    print(f"          f({c}) = {f(c)}")
    print(f"          → {classification}")
    print()

# Plot
x = np.linspace(-1, 5, 200)

plt.figure(figsize=(10, 6))
plt.plot(x, f(x), 'b-', linewidth=2, label='f(x) = x³ - 6x² + 9x + 2')

# Mark critical points
for c in critical_points:
    color = 'red' if f_double_prime(c) < 0 else 'green'
    label = 'Max' if f_double_prime(c) < 0 else 'Min'
    plt.scatter([c], [f(c)], color=color, s=100, zorder=5, label=f'Local {label} at x = {c}')

plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.title('Finding Extrema Using Derivatives', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## 7. Optimization: Maximum Economic Yield (MEY)

### The Problem

In fisheries management, **Maximum Economic Yield (MEY)** occurs at the stock level that maximizes profit, not just harvest.

### Profit Function

$$\pi(S) = pG(S) - cE$$

where:
- $\pi$ = profit
- $p$ = price per unit of fish
- $G(S)$ = sustainable yield at stock level $S$
- $c$ = cost per unit of effort
- $E$ = fishing effort

For sustainable fishing with catch rate proportional to effort and stock:
$$H = qES$$

where $q$ is the **catchability coefficient**.

```python
# MEY vs MSY Analysis

# Schaefer model parameters
g = 0.5        # Intrinsic growth rate
K = 1000       # Carrying capacity (tonnes)
p = 100        # Price per tonne ($)
c = 500        # Cost per unit effort ($)
q = 0.001      # Catchability coefficient

# Stock levels
S = np.linspace(100, K, 200)

# Sustainable yield: G(S) = gS(1 - S/K)
G = g * S * (1 - S/K)

# Revenue: R = p * G(S)
Revenue = p * G

# Effort required to maintain stock S: E = (g/q)(1 - S/K)
E = (g/q) * (1 - S/K)

# Cost: C = c * E
Cost = c * E

# Profit: π = Revenue - Cost
Profit = Revenue - Cost

# MSY: S* = K/2
S_MSY = K/2
G_MSY = g * K / 4

# MEY: Find maximum profit
idx_MEY = np.argmax(Profit)
S_MEY = S[idx_MEY]
Profit_MEY = Profit[idx_MEY]

# Plot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Left: Revenue, Cost, Profit
ax1.plot(S, Revenue, 'b-', linewidth=2, label='Revenue = p·G(S)')
ax1.plot(S, Cost, 'r-', linewidth=2, label='Cost = c·E')
ax1.plot(S, Profit, 'g-', linewidth=2, label='Profit = Revenue - Cost')
ax1.axvline(x=S_MSY, color='blue', linestyle=':', alpha=0.7, label=f'MSY stock = {S_MSY}')
ax1.axvline(x=S_MEY, color='green', linestyle='--', alpha=0.7, label=f'MEY stock ≈ {S_MEY:.0f}')
ax1.scatter([S_MEY], [Profit_MEY], color='green', s=100, zorder=5)
ax1.axhline(y=0, color='gray', linewidth=0.5)
ax1.set_xlabel('Stock S (tonnes)', fontsize=12)
ax1.set_ylabel('$ Value', fontsize=12)
ax1.set_title('Revenue, Cost, and Profit vs Stock Level', fontsize=14)
ax1.legend(loc='upper left', fontsize=9)
ax1.grid(True, alpha=0.3)

# Right: Comparison
ax2.bar(['MSY Stock', 'MEY Stock'], [S_MSY, S_MEY], color=['blue', 'green'], alpha=0.7)
ax2.set_ylabel('Stock Level (tonnes)', fontsize=12)
ax2.set_title('MSY vs MEY: Optimal Stock Levels', fontsize=14)

for i, (val, label) in enumerate(zip([S_MSY, S_MEY], ['MSY', 'MEY'])):
    ax2.text(i, val + 20, f'{val:.0f}', ha='center', fontsize=12)

plt.tight_layout()
plt.show()

print("="*60)
print("MSY vs MEY Comparison")
print("="*60)
print(f"Maximum Sustainable Yield (MSY):")
print(f"  Stock level: S_MSY = K/2 = {S_MSY} tonnes")
print(f"  Harvest: G_MSY = gK/4 = {G_MSY} tonnes/year")
print(f"\nMaximum Economic Yield (MEY):")
print(f"  Stock level: S_MEY ≈ {S_MEY:.0f} tonnes")
print(f"  Maximum profit: ${Profit_MEY:.2f}/year")
print(f"\nKey Insight: MEY typically occurs at a HIGHER stock level than MSY")
print(f"             because it's cheaper to catch fish when they're abundant!")
```

## 8. Optimization Problem Solving Strategy

### Step-by-Step Approach

1. **Understand the problem:** What quantity do we want to optimize?
2. **Define variables:** Assign symbols to all relevant quantities
3. **Write the objective function:** Express the quantity to optimize in terms of one variable
4. **Find the domain:** What values are physically meaningful?
5. **Find critical points:** Set the derivative equal to zero and solve
6. **Classify critical points:** Use first or second derivative test
7. **Check endpoints:** If domain is closed, compare values at endpoints
8. **Interpret the result:** Answer the original question with units

```python
# Example: Optimization word problem

print("PROBLEM: A rectangular fish pond has a fixed perimeter of 200 meters.")
print("What dimensions maximize the area of the pond?")
print()
print("="*60)
print("SOLUTION")
print("="*60)
print()
print("Step 1: Define variables")
print("  Let x = length and y = width")
print()
print("Step 2: Constraint")
print("  Perimeter: 2x + 2y = 200")
print("  So: y = 100 - x")
print()
print("Step 3: Objective function (Area)")
print("  A(x) = x · y = x(100 - x) = 100x - x²")
print()
print("Step 4: Domain")
print("  0 < x < 100 (both dimensions must be positive)")
print()
print("Step 5: Find critical points")
print("  A'(x) = 100 - 2x = 0")
print("  x = 50")
print()
print("Step 6: Classify (Second Derivative Test)")
print("  A''(x) = -2 < 0")
print("  → Local maximum at x = 50")
print()
print("Step 7: Find dimensions")
print("  x = 50 m, y = 100 - 50 = 50 m")
print()
print("Step 8: Interpret")
print("  Maximum area = 50 × 50 = 2500 m²")
print("  The optimal pond is a SQUARE!")

# Visualization
x = np.linspace(0.1, 99.9, 200)
A = x * (100 - x)

plt.figure(figsize=(10, 6))
plt.plot(x, A, 'b-', linewidth=2, label='Area A(x) = x(100 - x)')
plt.scatter([50], [2500], color='red', s=100, zorder=5, label='Maximum at (50, 2500)')
plt.axvline(x=50, color='red', linestyle='--', alpha=0.5)
plt.axhline(y=2500, color='red', linestyle='--', alpha=0.5)

plt.xlabel('Length x (m)', fontsize=12)
plt.ylabel('Area A (m²)', fontsize=12)
plt.title('Maximizing Area with Fixed Perimeter', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
