## 1. The Concept of a Limit

### Why Limits Matter

Limits are the foundation of calculus. They allow us to:
- Define instantaneous rates of change (derivatives)
- Calculate areas under curves (integrals)
- Analyze function behavior near problematic points

### Informal Definition

We write:

$$\lim_{x \to a} f(x) = L$$

and say "the limit of $f(x)$ as $x$ approaches $a$ equals $L$" if $f(x)$ gets arbitrarily close to $L$ as $x$ gets arbitrarily close to $a$.

> **Key Insight:** The limit describes what happens *near* $x = a$, not necessarily *at* $x = a$.

```python
# Visualizing a limit: f(x) = (x^2 - 1)/(x - 1) as x approaches 1

# This function is undefined at x = 1, but has a limit there
# f(x) = (x^2 - 1)/(x - 1) = (x+1)(x-1)/(x-1) = x + 1 for x ≠ 1

x_left = np.linspace(-1, 0.99, 100)
x_right = np.linspace(1.01, 3, 100)

def f(x):
    return (x**2 - 1) / (x - 1)

plt.figure(figsize=(10, 6))
plt.plot(x_left, f(x_left), 'b-', linewidth=2, label='f(x) = (x² - 1)/(x - 1)')
plt.plot(x_right, f(x_right), 'b-', linewidth=2)
plt.scatter([1], [2], facecolor='white', edgecolor='blue', s=100, zorder=5, label='Hole at (1, 2)')
plt.axhline(y=2, color='red', linestyle='--', alpha=0.5, label='Limit = 2')
plt.axvline(x=1, color='gray', linestyle=':', alpha=0.5)

# Show approach from both sides
plt.annotate('', xy=(1, 2), xytext=(0.6, 1.6),
             arrowprops=dict(arrowstyle='->', color='green', lw=2))
plt.annotate('', xy=(1, 2), xytext=(1.4, 2.4),
             arrowprops=dict(arrowstyle='->', color='green', lw=2))

plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.title('Limit of f(x) as x → 1 equals 2, even though f(1) is undefined', fontsize=14)
plt.legend(loc='lower right')
plt.grid(True, alpha=0.3)
plt.xlim(-1, 3)
plt.ylim(-1, 4)
plt.show()

# Numerical verification
print("Approaching x = 1 from the left:")
for x in [0.9, 0.99, 0.999, 0.9999]:
    print(f"  f({x}) = {f(x):.6f}")

print("\nApproaching x = 1 from the right:")
for x in [1.1, 1.01, 1.001, 1.0001]:
    print(f"  f({x}) = {f(x):.6f}")
```

## 2. One-Sided Limits

### Left-Hand and Right-Hand Limits

- **Left-hand limit:** $\displaystyle\lim_{x \to a^-} f(x)$ — approach $a$ from values less than $a$
- **Right-hand limit:** $\displaystyle\lim_{x \to a^+} f(x)$ — approach $a$ from values greater than $a$

**The two-sided limit exists if and only if both one-sided limits exist and are equal:**

$$\lim_{x \to a} f(x) = L \iff \lim_{x \to a^-} f(x) = \lim_{x \to a^+} f(x) = L$$

```python
# Example: One-sided limits with a step function

def step_function(x):
    return np.where(x < 2, x**2, 3*x - 2)

x_left = np.linspace(0, 1.99, 100)
x_right = np.linspace(2.01, 4, 100)

plt.figure(figsize=(10, 6))
plt.plot(x_left, x_left**2, 'b-', linewidth=2, label='f(x) = x² for x < 2')
plt.plot(x_right, 3*x_right - 2, 'r-', linewidth=2, label='f(x) = 3x - 2 for x ≥ 2')

# Show the one-sided limits
plt.scatter([2], [4], facecolor='white', edgecolor='blue', s=100, zorder=5, label='lim⁻ = 4')
plt.scatter([2], [4], facecolor='red', edgecolor='red', s=100, zorder=5, label='lim⁺ = 4')

plt.axvline(x=2, color='gray', linestyle=':', alpha=0.5)

plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.title('Both one-sided limits equal 4, so the two-sided limit exists', fontsize=14)
plt.legend(loc='upper left')
plt.grid(True, alpha=0.3)
plt.show()

print("Left-hand limit as x → 2⁻: lim(x²) = 4")
print("Right-hand limit as x → 2⁺: lim(3x - 2) = 4")
print("Since both equal 4, the limit exists: lim f(x) as x → 2 = 4")
```

## 3. Limit Laws

For limits that exist, we can apply these algebraic rules:

| Law | Formula |
|-----|--------|
| Sum | $\displaystyle\lim_{x \to a} [f(x) + g(x)] = \lim_{x \to a} f(x) + \lim_{x \to a} g(x)$ |
| Difference | $\displaystyle\lim_{x \to a} [f(x) - g(x)] = \lim_{x \to a} f(x) - \lim_{x \to a} g(x)$ |
| Product | $\displaystyle\lim_{x \to a} [f(x) \cdot g(x)] = \lim_{x \to a} f(x) \cdot \lim_{x \to a} g(x)$ |
| Quotient | $\displaystyle\lim_{x \to a} \frac{f(x)}{g(x)} = \frac{\lim_{x \to a} f(x)}{\lim_{x \to a} g(x)}$ (if denominator ≠ 0) |
| Power | $\displaystyle\lim_{x \to a} [f(x)]^n = \left[\lim_{x \to a} f(x)\right]^n$ |

```python
# Example: Evaluating limits using limit laws

# Find lim (x² + 3x - 2) as x → 2
# Using direct substitution (since polynomial is continuous)

def f(x):
    return x**2 + 3*x - 2

a = 2

# Step-by-step calculation
print("Find lim(x² + 3x - 2) as x → 2")
print("="*40)
print("\nUsing limit laws:")
print(f"  lim(x²) = {a}² = {a**2}")
print(f"  lim(3x) = 3·{a} = {3*a}")
print(f"  lim(-2) = -2")
print(f"\nSum law:")
print(f"  lim(x² + 3x - 2) = {a**2} + {3*a} + (-2) = {f(a)}")
print(f"\nVerification: f({a}) = {f(a)}")
```

## 4. Continuity

### Definition

A function $f$ is **continuous at $x = a$** if three conditions are met:

1. $f(a)$ is defined (the function exists at the point)
2. $\displaystyle\lim_{x \to a} f(x)$ exists (the limit exists)
3. $\displaystyle\lim_{x \to a} f(x) = f(a)$ (the limit equals the function value)

### Types of Discontinuities

| Type | Description | Can be "fixed"? |
|------|-------------|-----------------|
| **Removable** | Limit exists but $f(a)$ undefined or wrong | Yes, redefine $f(a)$ |
| **Jump** | Left and right limits differ | No |
| **Infinite** | Function goes to ±∞ | No |

```python
# Visualizing the three types of discontinuity

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# 1. Removable discontinuity
x1 = np.linspace(-2, 2, 200)
x1_left = x1[x1 < 0]
x1_right = x1[x1 > 0]
axes[0].plot(x1_left, np.sin(x1_left)/x1_left, 'b-', lw=2)
axes[0].plot(x1_right, np.sin(x1_right)/x1_right, 'b-', lw=2)
axes[0].scatter([0], [1], facecolor='white', edgecolor='blue', s=100, zorder=5)
axes[0].set_title('Removable Discontinuity\nf(x) = sin(x)/x', fontsize=12)
axes[0].set_xlabel('x')
axes[0].set_ylabel('f(x)')
axes[0].grid(True, alpha=0.3)
axes[0].set_ylim(-0.5, 1.5)

# 2. Jump discontinuity
x2_left = np.linspace(-2, 0.99, 100)
x2_right = np.linspace(1.01, 3, 100)
axes[1].plot(x2_left, x2_left**2, 'b-', lw=2)
axes[1].plot(x2_right, x2_right + 2, 'b-', lw=2)
axes[1].scatter([1], [1], facecolor='blue', edgecolor='blue', s=100, zorder=5)
axes[1].scatter([1], [3], facecolor='white', edgecolor='blue', s=100, zorder=5)
axes[1].set_title('Jump Discontinuity\nLeft limit ≠ Right limit', fontsize=12)
axes[1].set_xlabel('x')
axes[1].set_ylabel('f(x)')
axes[1].grid(True, alpha=0.3)

# 3. Infinite discontinuity
x3 = np.linspace(-2, 2, 400)
x3 = x3[np.abs(x3) > 0.1]  # Avoid division by zero
axes[2].plot(x3, 1/x3, 'b-', lw=2)
axes[2].axvline(x=0, color='red', linestyle='--', alpha=0.5)
axes[2].set_title('Infinite Discontinuity\nf(x) = 1/x', fontsize=12)
axes[2].set_xlabel('x')
axes[2].set_ylabel('f(x)')
axes[2].grid(True, alpha=0.3)
axes[2].set_ylim(-10, 10)

plt.tight_layout()
plt.show()
```

## 5. Average Rate of Change

### From Position to Velocity

If $s(t)$ gives position at time $t$, the **average velocity** over the interval $[a, b]$ is:

$$v_{avg} = \frac{\Delta s}{\Delta t} = \frac{s(b) - s(a)}{b - a}$$

**Geometrically:** This is the slope of the **secant line** connecting $(a, s(a))$ and $(b, s(b))$.

```python
# Example: Average rate of change (secant line)

# Position function: s(t) = t² (dropped object)
def s(t):
    return t**2

# Points for secant line
a, b = 1, 3

# Calculate average velocity
avg_velocity = (s(b) - s(a)) / (b - a)

# Create plot
t = np.linspace(0, 4, 100)

plt.figure(figsize=(10, 6))
plt.plot(t, s(t), 'b-', linewidth=2, label='s(t) = t²')

# Secant line: y - s(a) = m(x - a)
secant_t = np.linspace(0, 4, 100)
secant_s = s(a) + avg_velocity * (secant_t - a)
plt.plot(secant_t, secant_s, 'r--', linewidth=2, label=f'Secant line (slope = {avg_velocity})')

# Mark points
plt.scatter([a, b], [s(a), s(b)], color='red', s=100, zorder=5)
plt.annotate(f'({a}, {s(a)})', (a, s(a)), xytext=(a-0.5, s(a)+1), fontsize=10)
plt.annotate(f'({b}, {s(b)})', (b, s(b)), xytext=(b+0.1, s(b)+1), fontsize=10)

plt.xlabel('Time t (seconds)', fontsize=12)
plt.ylabel('Position s(t) (meters)', fontsize=12)
plt.title('Average Velocity = Slope of Secant Line', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.xlim(0, 4)
plt.ylim(0, 16)
plt.show()

print(f"Position function: s(t) = t²")
print(f"Average velocity from t = {a} to t = {b}:")
print(f"  v_avg = [s({b}) - s({a})] / [{b} - {a}]")
print(f"        = [{s(b)} - {s(a)}] / [{b-a}]")
print(f"        = {s(b)-s(a)} / {b-a}")
print(f"        = {avg_velocity} m/s")
```

## 6. Instantaneous Rate of Change: The Derivative

### The Big Idea

What if we want the velocity at a **single instant**, not over an interval?

**Strategy:** Make the interval smaller and smaller, taking the limit as $b \to a$.

### Definition of the Derivative

The **derivative** of $f$ at $x = a$, denoted $f'(a)$, is:

$$f'(a) = \lim_{h \to 0} \frac{f(a + h) - f(a)}{h}$$

or equivalently:

$$f'(a) = \lim_{x \to a} \frac{f(x) - f(a)}{x - a}$$

**Geometrically:** The derivative is the slope of the **tangent line** at $(a, f(a))$.

```python
# Visualizing secant lines approaching the tangent line

def f(x):
    return x**2

a = 2  # Point of tangency

# The derivative at x = 2 (for f(x) = x²)
# f'(x) = 2x, so f'(2) = 4
slope_tangent = 2 * a

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

x = np.linspace(0, 4, 200)

# Left plot: Secant lines approaching tangent
axes[0].plot(x, f(x), 'b-', linewidth=2, label='f(x) = x²')

# Secant lines for different values of h
colors = ['orange', 'purple', 'green', 'red']
h_values = [1.5, 1.0, 0.5, 0.1]

for h, color in zip(h_values, colors):
    slope = (f(a + h) - f(a)) / h
    secant_y = f(a) + slope * (x - a)
    axes[0].plot(x, secant_y, '--', color=color, alpha=0.7, label=f'h = {h}, slope = {slope:.2f}')
    axes[0].scatter([a + h], [f(a + h)], color=color, s=50, zorder=5)

# Tangent line
tangent_y = f(a) + slope_tangent * (x - a)
axes[0].plot(x, tangent_y, 'k-', linewidth=2, label=f'Tangent (slope = {slope_tangent})')
axes[0].scatter([a], [f(a)], color='black', s=100, zorder=5)

axes[0].set_xlabel('x', fontsize=12)
axes[0].set_ylabel('f(x)', fontsize=12)
axes[0].set_title('Secant Lines → Tangent Line as h → 0', fontsize=14)
axes[0].legend(loc='upper left', fontsize=9)
axes[0].grid(True, alpha=0.3)
axes[0].set_xlim(0, 4)
axes[0].set_ylim(0, 16)

# Right plot: Show how slope converges
h_array = np.linspace(0.01, 2, 100)
slopes = (f(a + h_array) - f(a)) / h_array

axes[1].plot(h_array, slopes, 'b-', linewidth=2, label='Secant slope')
axes[1].axhline(y=slope_tangent, color='red', linestyle='--', linewidth=2, label=f'Derivative = {slope_tangent}')
axes[1].set_xlabel('h', fontsize=12)
axes[1].set_ylabel('Slope of secant line', fontsize=12)
axes[1].set_title('Secant Slope Converges to Derivative as h → 0', fontsize=14)
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"For f(x) = x² at x = {a}:")
print(f"  f'({a}) = lim (f({a}+h) - f({a})) / h as h → 0")
print(f"         = lim (({a}+h)² - {a}²) / h as h → 0")
print(f"         = lim ({a}² + 2·{a}·h + h² - {a}²) / h as h → 0")
print(f"         = lim (2·{a}·h + h²) / h as h → 0")
print(f"         = lim (2·{a} + h) as h → 0")
print(f"         = {slope_tangent}")
```

## 7. Finding Derivatives from First Principles

### The Process

To find $f'(x)$ using the limit definition:

1. Write the difference quotient: $\displaystyle\frac{f(x + h) - f(x)}{h}$
2. Simplify algebraically (often by factoring or expanding)
3. Cancel the $h$ in the denominator
4. Take the limit as $h \to 0$

```python
# Example: Derivative of f(x) = x³ from first principles

print("Find the derivative of f(x) = x³ using first principles")
print("="*55)
print()
print("Step 1: Write the difference quotient")
print("  [f(x + h) - f(x)] / h = [(x + h)³ - x³] / h")
print()
print("Step 2: Expand (x + h)³")
print("  (x + h)³ = x³ + 3x²h + 3xh² + h³")
print()
print("Step 3: Substitute and simplify")
print("  [(x³ + 3x²h + 3xh² + h³) - x³] / h")
print("  = [3x²h + 3xh² + h³] / h")
print("  = h[3x² + 3xh + h²] / h")
print("  = 3x² + 3xh + h²")
print()
print("Step 4: Take the limit as h → 0")
print("  f'(x) = lim (3x² + 3xh + h²) as h → 0")
print("        = 3x² + 0 + 0")
print("        = 3x²")
print()
print("Therefore: d/dx(x³) = 3x²")
```

## 8. The Power Rule

### The Most Important Differentiation Rule

For any real number $n$:

$$\frac{d}{dx}(x^n) = n \cdot x^{n-1}$$

**Examples:**

| Function | Derivative |
|----------|------------|
| $f(x) = x^5$ | $f'(x) = 5x^4$ |
| $f(x) = x^2$ | $f'(x) = 2x$ |
| $f(x) = x$ | $f'(x) = 1$ |
| $f(x) = x^{-1} = \frac{1}{x}$ | $f'(x) = -x^{-2} = -\frac{1}{x^2}$ |
| $f(x) = \sqrt{x} = x^{1/2}$ | $f'(x) = \frac{1}{2}x^{-1/2} = \frac{1}{2\sqrt{x}}$ |

### Constant Multiple Rule

$$\frac{d}{dx}[c \cdot f(x)] = c \cdot f'(x)$$

### Sum/Difference Rule

$$\frac{d}{dx}[f(x) \pm g(x)] = f'(x) \pm g'(x)$$

```python
# Applying the power rule to polynomials

print("Example: Find the derivative of f(x) = 4x³ - 2x² + 5x - 7")
print("="*55)
print()
print("Apply the power rule term by term:")
print()
print("  d/dx(4x³) = 4 · 3x² = 12x²")
print("  d/dx(-2x²) = -2 · 2x = -4x")
print("  d/dx(5x) = 5 · 1 = 5")
print("  d/dx(-7) = 0  (constant rule)")
print()
print("Therefore:")
print("  f'(x) = 12x² - 4x + 5")
print()

# Verify numerically
def f(x):
    return 4*x**3 - 2*x**2 + 5*x - 7

def f_prime(x):
    return 12*x**2 - 4*x + 5

# Compare derivative with numerical approximation
x_test = 2
h = 0.0001
numerical_deriv = (f(x_test + h) - f(x_test)) / h

print(f"Verification at x = {x_test}:")
print(f"  f'({x_test}) using formula: {f_prime(x_test)}")
print(f"  Numerical approximation: {numerical_deriv:.4f}")
```

## 9. Tangent Lines

### Equation of the Tangent Line

The tangent line to $y = f(x)$ at the point $(a, f(a))$ has:
- **Slope:** $m = f'(a)$
- **Equation:** $y - f(a) = f'(a)(x - a)$

or in slope-intercept form:

$$y = f'(a) \cdot x + [f(a) - a \cdot f'(a)]$$

```python
# Example: Finding and plotting a tangent line

def f(x):
    return x**3 - 3*x + 1

def f_prime(x):
    return 3*x**2 - 3

# Find tangent line at x = 2
a = 2
slope = f_prime(a)
y_point = f(a)

# Tangent line equation: y - f(a) = f'(a)(x - a)
# y = f'(a) * x + (f(a) - a * f'(a))
intercept = y_point - slope * a

# Plot
x = np.linspace(-1, 3, 200)
tangent_y = slope * x + intercept

plt.figure(figsize=(10, 6))
plt.plot(x, f(x), 'b-', linewidth=2, label='f(x) = x³ - 3x + 1')
plt.plot(x, tangent_y, 'r--', linewidth=2, label=f'Tangent at x = {a}')
plt.scatter([a], [y_point], color='red', s=100, zorder=5)

plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title(f'Tangent Line to f(x) = x³ - 3x + 1 at x = {a}', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.xlim(-1, 3)
plt.ylim(-5, 10)
plt.show()

print(f"For f(x) = x³ - 3x + 1 at x = {a}:")
print(f"  Point of tangency: ({a}, {y_point})")
print(f"  f'(x) = 3x² - 3")
print(f"  f'({a}) = 3({a})² - 3 = {slope}")
print(f"\nTangent line equation:")
print(f"  y - {y_point} = {slope}(x - {a})")
print(f"  y = {slope}x + {intercept}")
```
