## Part A: Exponential Models (~25 min)

The general exponential model is:
$$P(t) = P_0 \, e^{kt}, \qquad k > 0 \text{ (growth)}, \quad k < 0 \text{ (decay)}$$

Equivalent base-$b$ form: $P(t) = P_0 \, b^t$ where $b = e^k$.

````python
# A.1 — Plastic production: 8.4% annual growth, P0 = 2 M tonnes in 1950
P0_plastic = 2          # million tonnes
growth_rate = 0.084     # 8.4% per year

k_plastic = np.log(1 + growth_rate)   # k = ln(1.084)
print(f"k = ln(1.084) = {k_plastic:.4f} per year")

# Predict 2025 production
t_2025 = 2025 - 1950
P_2025 = P0_plastic * np.exp(k_plastic * t_2025)
print(f"Predicted production in 2025: {P_2025:.1f} million tonnes")
print(f"(Actual ~500 M tonnes — how does the model compare?)")
````

````python
# A.2 — Bacterial growth: N0 = 1000 cells, doubles every 20 minutes
N0 = 1000
T_double = 20           # minutes

# Both forms — verify they agree
k_bact = np.log(2) / T_double

def N_v1(t): return N0 * 2**(t / T_double)
def N_v2(t): return N0 * np.exp(k_bact * t)

print(f"k = ln(2)/20 = {k_bact:.4f} per minute")
print()
for t_hrs in [1, 2, 3]:
    t = t_hrs * 60
    print(f"t = {t_hrs}h: 2^(t/T) = {N_v1(t):>12,.0f}  |  e^(kt) = {N_v2(t):>12,.0f}")
````

````python
# A.3 — Carbon-14 decay: A0 = 100 units, half-life = 5,730 years
A0_c14 = 100
T_half = 5730           # years

lambda_c14 = np.log(2) / T_half    # decay constant (positive)

def A_c14(t):
    return A0_c14 * np.exp(-lambda_c14 * t)   # note: negative exponent

print(f"λ = ln(2)/5730 = {lambda_c14:.6f} per year")
print()
for n_halves in [1, 2, 3]:
    t = n_halves * T_half
    print(f"After {n_halves} half-life(s) ({t:,} years): {A_c14(t):.2f} units  ({A_c14(t):.1f}%)")
````

## Part B: Doubling Time & the Rule of 70 (~15 min)

**Exact formula:**
$$T_{\text{double}} = \frac{\ln 2}{k}$$

**Rule of 70 approximation** (for percentage rates $r$):
$$T_{\text{double}} \approx \frac{70}{r}$$

````python
# B.1 — Compare exact formula vs Rule of 70 for plastic growth
r_pct = 8.4

T_exact  = np.log(2) / k_plastic
T_rule70 = 70 / r_pct

print(f"Exact doubling time:      {T_exact:.2f} years")
print(f"Rule of 70 approximation: {T_rule70:.2f} years")
print(f"Error: {abs(T_exact - T_rule70)/T_exact*100:.1f}%")

# Accuracy across a range of growth rates
print()
print(f"{'Rate':>8} | {'Exact':>10} | {'Rule 70':>10} | {'Error':>7}")
print("-" * 42)
for r in [1, 2, 4, 8.4, 10, 15, 20]:
    T_e = np.log(2) / np.log(1 + r/100)
    T_r = 70 / r
    print(f"{r:>7.1f}% | {T_e:>9.2f}y | {T_r:>9.2f}y | {abs(T_e-T_r)/T_e*100:>6.1f}%")
````

````python
# B.2 — When does plastic production hit 1 billion and 10 billion tonnes?
def years_to_reach(target_M_tonnes):
    t = np.log(target_M_tonnes / P0_plastic) / k_plastic
    return t, 1950 + t

t_1b,  yr_1b  = years_to_reach(1_000)    # 1 billion tonnes
t_10b, yr_10b = years_to_reach(10_000)   # 10 billion tonnes

print(f"1 billion tonnes:  year {yr_1b:.0f}  (t = {t_1b:.0f} yrs from 1950)")
print(f"10 billion tonnes: year {yr_10b:.0f}  (t = {t_10b:.0f} yrs from 1950)")
print()
print("Note: these projections assume constant 8.4% growth — discuss with your group")
print("whether that assumption is realistic.")
````

## Part C: Logarithmic Solving (~15 min)

To find **when** an exponential reaches a target value $C$:
$$P_0 e^{kt} = C \implies t = \frac{\ln(C/P_0)}{k}$$

````python
# C.1 — Invasive carp: P(t) = 50 * e^(0.15t), t in months
P0_carp = 50
k_carp  = 0.15
limit   = 1000    # eradication impossible above this

def carp(t): return P0_carp * np.exp(k_carp * t)
def t_to_reach(target): return np.log(target / P0_carp) / k_carp

for target in [500, 1000, 5000]:
    t = t_to_reach(target)
    print(f"Reach {target:>5}: {t:>5.1f} months  ({t/12:.1f} years)")

# Visualise
t_vals = np.linspace(0, 55, 300)
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(t_vals, carp(t_vals), "steelblue", lw=2)
ax.axhline(limit, color="red", ls="--", label=f"Eradication limit ({limit} fish)")
ax.set_xlabel("Time (months)"); ax.set_ylabel("Carp population")
ax.set_title("Invasive Carp Population Growth")
ax.legend(); plt.tight_layout(); plt.show()
````

````python
# C.2 — Carbon dating: fossil has 25% of original C-14 remaining
fraction = 0.25

# Solve: e^(-lambda * t) = 0.25  =>  t = -ln(0.25) / lambda
t_fossil = -np.log(fraction) / lambda_c14
print(f"A fossil with 25% C-14 remaining is ~{t_fossil:.0f} years old")
print(f"That is {t_fossil/T_half:.2f} half-lives")

# GROUP EXERCISE — try your own fraction:
# my_fraction = 0.10   # e.g. 10%
# my_age = -np.log(my_fraction) / lambda_c14
# print(f"At {my_fraction*100:.0f}% remaining: {my_age:.0f} years")
````
