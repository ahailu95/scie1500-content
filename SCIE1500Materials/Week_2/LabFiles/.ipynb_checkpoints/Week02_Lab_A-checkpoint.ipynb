## 📋 Scenario

You are a consultant for **Royal Perth Hospital's** infection control team. A strain of antibiotic-resistant *Staphylococcus aureus* (MRSA) has been detected on a surgical instrument.

- Doubling time at room temperature: **25 minutes**
- Initial contamination: **800 cells**
- **Hospital policy:** If count exceeds **1,000,000 cells**, the surgical suite shuts down for sterilization — cost: $50,000 + 48-hour delay.

---
## 🎯 Your Task

| Part | Topic | Time |
|------|-------|------|
| A | Build growth models in two forms; find shutdown time | ~20 min |
| B | Effect of refrigeration (4°C, T = 180 min) | ~15 min |
| C | Sterilization decision — policy recommendation | ~10 min |
| D | Verify the Rule of 70 for this scenario | ~10 min |

````python
# Setup — run first
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams.update({'figure.dpi': 120, 'font.size': 11})

N0        = 800
T_room    = 25          # doubling time at room temp (minutes)
threshold = 1_000_000   # shutdown threshold

k_room = np.log(2) / T_room
print(f"Room-temp growth constant k = {k_room:.4f}/min")
print(f"Equivalent: {(np.exp(k_room)-1)*100:.3f}% growth per minute")
````

---
## Part A: Modeling Bacterial Growth (~20 min)

````python
# A.1 — Two equivalent forms; verify they agree
def N_v1(t): return N0 * 2**(t / T_room)              # doubling form
def N_v2(t): return N0 * np.exp(k_room * t)           # e-form

print("Both forms should agree:")
for t_h in [1, 2, 3]:
    t = t_h * 60
    print(f"  t={t_h}h ({t:3d} min): 2^(t/T) = {N_v1(t):>12,.0f}   e^(kt) = {N_v2(t):>12,.0f}")
````

````python
# A.2 — Time to reach shutdown threshold
# Solve: 800 * 2^(t/25) = 1,000,000
t_shutdown = T_room * np.log2(threshold / N0)
print(f"Time to reach {threshold:,} cells: {t_shutdown:.1f} min = {t_shutdown/60:.2f} hours")

t_vals = np.linspace(0, 350, 600)
fig, ax = plt.subplots(figsize=(9, 4))
ax.plot(t_vals, N_v1(t_vals), "steelblue", lw=2, label="MRSA growth (room temp)")
ax.axhline(threshold, color="red",    ls="--", lw=1.5, label="Shutdown threshold (1M cells)")
ax.axvline(t_shutdown, color="orange", ls=":",  lw=1.5, label=f"Threshold at {t_shutdown:.0f} min")
ax.set_yscale("log")
ax.set_xlabel("Time (minutes)"); ax.set_ylabel("Bacterial count (log scale)")
ax.set_title("MRSA Growth — Royal Perth Hospital")
ax.legend(); plt.tight_layout(); plt.show()
````

## Part B: Effect of Refrigeration (~15 min)

At **4°C**, doubling time increases to **180 minutes**.

````python
# B.1 — Cold storage model
T_cold  = 180
k_cold  = np.log(2) / T_cold

def N_cold(t): return N0 * np.exp(k_cold * t)

t_shutdown_cold = T_cold * np.log2(threshold / N0)
print(f"Room temp: shutdown at {t_shutdown:.1f} min ({t_shutdown/60:.2f} h)")
print(f"At 4°C:    shutdown at {t_shutdown_cold:.0f} min ({t_shutdown_cold/60:.1f} h)")
print(f"Refrigeration buys {t_shutdown_cold/t_shutdown:.1f}× more time")

fig, ax = plt.subplots(figsize=(9, 5))
ax.plot(t_vals, N_v1(t_vals),  "steelblue",  lw=2, label=f"Room temp (T={T_room} min)")
ax.plot(t_vals, N_cold(t_vals), "seagreen",   lw=2, label=f"4°C refrigerated (T={T_cold} min)")
ax.axhline(threshold, color="red", ls="--", label="Shutdown threshold")
ax.set_yscale("log")
ax.set_xlabel("Time (minutes)"); ax.set_ylabel("Bacterial count (log scale)")
ax.set_title("MRSA Growth: Room Temp vs Refrigeration")
ax.legend(); plt.tight_layout(); plt.show()
````

## Part C: Sterilization Decision (~10 min)

- **Option 1:** Immediate sterilization — \$50,000 + 48-hour delay  
- **Option 2:** Refrigerate 12 hours, re-test, then decide

````python
# C.1 — What is the count after 12 hours of refrigeration?
t_12h = 12 * 60    # minutes
N_after = N_cold(t_12h)

print(f"Bacterial count after 12h fridge: {N_after:,.0f} cells")
print(f"Still below threshold ({threshold:,})? {N_after < threshold}")
print(f"Safety margin: {threshold/N_after:.1f}× below threshold")
````

## Part D: Rule of 70 Verification (~10 min)

````python
# D.1 — How accurate is Rule of 70 for this scenario?
r_exact_pct = (np.exp(k_room) - 1) * 100     # exact % growth rate per minute

T_rule70 = 70 / r_exact_pct
print(f"Exact % growth per minute:  {r_exact_pct:.4f}%")
print(f"Rule of 70 doubling time:   {T_rule70:.2f} min")
print(f"Exact doubling time:        {T_room} min")
print(f"Error: {abs(T_rule70 - T_room)/T_room*100:.1f}%")
print()
print("Note: Rule of 70 works best for small % rates; bacteria growth per minute is large.")
````
