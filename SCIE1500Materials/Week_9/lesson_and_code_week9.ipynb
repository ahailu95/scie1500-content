## 1. Basic Probability Concepts

### Sample Spaces and Events

- **Experiment:** A process with uncertain outcome
- **Sample Space ($S$):** Set of all possible outcomes
- **Event ($A$):** A subset of the sample space

### Definition of Probability

For equally likely outcomes:

$$P(A) = \frac{\text{Number of outcomes in } A}{\text{Total number of outcomes}} = \frac{|A|}{|S|}$$

### Properties

1. $0 \leq P(A) \leq 1$ for any event $A$
2. $P(S) = 1$ (something must happen)
3. $P(\emptyset) = 0$ (impossible event)

```python
# Example: Rolling a die

# Sample space
S = {1, 2, 3, 4, 5, 6}
print("Sample space S =", S)
print(f"|S| = {len(S)}")
print()

# Event A: Rolling an even number
A = {2, 4, 6}
print("Event A (even) =", A)
P_A = len(A) / len(S)
print(f"P(A) = {len(A)}/{len(S)} = {P_A:.4f}")
print()

# Event B: Rolling greater than 4
B = {5, 6}
print("Event B (>4) =", B)
P_B = len(B) / len(S)
print(f"P(B) = {len(B)}/{len(S)} = {P_B:.4f}")
```

## 2. Complement Rule

### Definition

The **complement** of event $A$, denoted $A'$ or $\bar{A}$, consists of all outcomes NOT in $A$.

### Formula

$$P(A') = 1 - P(A)$$

This is extremely useful: sometimes it's easier to calculate "not $A$" than $A$ directly.

```python
# Example: Probability of getting at least one head in 3 coin flips

# Direct method: count outcomes with at least one head
# Sample space for 3 flips: HHH, HHT, HTH, HTT, THH, THT, TTH, TTT
S = ['HHH', 'HHT', 'HTH', 'HTT', 'THH', 'THT', 'TTH', 'TTT']

# At least one head (complement of all tails)
all_tails = ['TTT']

P_no_heads = len(all_tails) / len(S)
P_at_least_one_head = 1 - P_no_heads

print("Sample space:", S)
print(f"Total outcomes: {len(S)}")
print()
print("Using complement rule:")
print(f"P(no heads) = 1/8 = {P_no_heads:.4f}")
print(f"P(at least one head) = 1 - P(no heads) = 1 - 1/8 = {P_at_least_one_head:.4f}")
print()

# Alternative calculation
P_alt = (0.5)**3
print(f"Alternative: P(no heads) = (1/2)³ = {P_alt:.4f}")
```

## 3. Addition Rule

### General Addition Rule

For any events $A$ and $B$:

$$P(A \cup B) = P(A) + P(B) - P(A \cap B)$$

**Why subtract?** To avoid double-counting outcomes in both $A$ and $B$.

### Mutually Exclusive Events

If $A$ and $B$ cannot both occur ($A \cap B = \emptyset$):

$$P(A \cup B) = P(A) + P(B)$$

```python
# Visualizing the addition rule with Venn diagrams

from matplotlib.patches import Circle

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Left: General case (overlapping)
circle1 = Circle((0.35, 0.5), 0.3, color='blue', alpha=0.3, label='A')
circle2 = Circle((0.65, 0.5), 0.3, color='red', alpha=0.3, label='B')
ax1.add_patch(circle1)
ax1.add_patch(circle2)
ax1.text(0.25, 0.5, 'A only', ha='center', va='center', fontsize=12)
ax1.text(0.75, 0.5, 'B only', ha='center', va='center', fontsize=12)
ax1.text(0.5, 0.5, 'A∩B', ha='center', va='center', fontsize=12)
ax1.set_xlim(0, 1)
ax1.set_ylim(0, 1)
ax1.set_aspect('equal')
ax1.set_title('General: P(A∪B) = P(A) + P(B) - P(A∩B)', fontsize=12)
ax1.axis('off')

# Right: Mutually exclusive
circle3 = Circle((0.3, 0.5), 0.2, color='blue', alpha=0.3, label='A')
circle4 = Circle((0.7, 0.5), 0.2, color='red', alpha=0.3, label='B')
ax2.add_patch(circle3)
ax2.add_patch(circle4)
ax2.text(0.3, 0.5, 'A', ha='center', va='center', fontsize=12)
ax2.text(0.7, 0.5, 'B', ha='center', va='center', fontsize=12)
ax2.set_xlim(0, 1)
ax2.set_ylim(0, 1)
ax2.set_aspect('equal')
ax2.set_title('Mutually Exclusive: P(A∪B) = P(A) + P(B)', fontsize=12)
ax2.axis('off')

plt.tight_layout()
plt.show()

# Numerical example
print("Example: Drawing a card from a standard deck")
print("A = Heart, B = Face card (J, Q, K)")
print()
P_A = 13/52  # Hearts
P_B = 12/52  # Face cards (4 suits × 3)
P_A_and_B = 3/52  # Heart face cards (J♥, Q♥, K♥)
P_A_or_B = P_A + P_B - P_A_and_B

print(f"P(Heart) = 13/52 = {P_A:.4f}")
print(f"P(Face) = 12/52 = {P_B:.4f}")
print(f"P(Heart AND Face) = 3/52 = {P_A_and_B:.4f}")
print(f"P(Heart OR Face) = {P_A:.4f} + {P_B:.4f} - {P_A_and_B:.4f} = {P_A_or_B:.4f}")
```

## 4. Conditional Probability

### Definition

The probability of $A$ **given that** $B$ has occurred:

$$P(A|B) = \frac{P(A \cap B)}{P(B)}$$

**Read as:** "Probability of A given B"

### Interpretation

Conditional probability restricts our attention to outcomes where $B$ occurred, then asks: what fraction of those are also in $A$?

```python
# Example: Medical testing

print("Medical Test Example")
print("="*50)
print()
print("Suppose:")
print("  1% of population has a disease (D)")
print("  Test is 90% accurate for sick people: P(+|D) = 0.90")
print("  Test is 95% accurate for healthy people: P(-|D') = 0.95")
print()

# Given information
P_D = 0.01           # P(Disease)
P_D_prime = 0.99     # P(No Disease)
P_pos_given_D = 0.90       # Sensitivity
P_neg_given_D_prime = 0.95 # Specificity
P_pos_given_D_prime = 0.05 # False positive rate

# Calculate joint probabilities
P_pos_and_D = P_pos_given_D * P_D
P_pos_and_D_prime = P_pos_given_D_prime * P_D_prime

# Total probability of positive test
P_pos = P_pos_and_D + P_pos_and_D_prime

print("Joint probabilities:")
print(f"  P(+ and D) = P(+|D)·P(D) = 0.90×0.01 = {P_pos_and_D:.4f}")
print(f"  P(+ and D') = P(+|D')·P(D') = 0.05×0.99 = {P_pos_and_D_prime:.4f}")
print(f"  P(+) = {P_pos:.4f}")
```

## 5. Bayes' Theorem

### The Formula

$$P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}$$

### Using Total Probability in the Denominator

$$P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B|A) \cdot P(A) + P(B|A') \cdot P(A')}$$

### The Key Insight

Bayes' theorem lets us **reverse** conditional probabilities. If we know $P(B|A)$ but want $P(A|B)$, Bayes' theorem is the tool.

```python
# Bayes' Theorem: What's the probability of disease given a positive test?

print("Bayes' Theorem Application")
print("="*50)
print()
print("Question: If you test positive, what's the probability you have the disease?")
print()

# P(D|+) = P(+|D) × P(D) / P(+)
P_D_given_pos = (P_pos_given_D * P_D) / P_pos

print("Applying Bayes' Theorem:")
print()
print("         P(+|D) × P(D)")
print("P(D|+) = ─────────────────────────────────")
print("         P(+|D)×P(D) + P(+|D')×P(D')")
print()
print(f"         0.90 × 0.01")
print(f"       = ───────────────────────")
print(f"         0.90×0.01 + 0.05×0.99")
print()
print(f"         {P_pos_and_D:.4f}")
print(f"       = ────────")
print(f"         {P_pos:.4f}")
print()
print(f"       = {P_D_given_pos:.4f} = {P_D_given_pos*100:.1f}%")
print()
print("Surprising! Even with a positive test, there's only ~15% chance of disease.")
print("This is because the disease is rare (base rate fallacy).")
```

```python
# Visualization: Base rate effect

base_rates = np.linspace(0.001, 0.5, 100)
P_D_given_pos_array = []

for p_d in base_rates:
    p_pos = P_pos_given_D * p_d + P_pos_given_D_prime * (1 - p_d)
    p_d_pos = (P_pos_given_D * p_d) / p_pos
    P_D_given_pos_array.append(p_d_pos)

plt.figure(figsize=(10, 6))
plt.plot(base_rates * 100, np.array(P_D_given_pos_array) * 100, 'b-', linewidth=2)
plt.axhline(y=50, color='gray', linestyle='--', alpha=0.5)
plt.scatter([1], [P_D_given_pos * 100], color='red', s=100, zorder=5, label=f'Our example: {P_D_given_pos*100:.1f}%')

plt.xlabel('Base rate P(D) (%)', fontsize=12)
plt.ylabel('P(D|+) (%)', fontsize=12)
plt.title('Effect of Base Rate on Posterior Probability', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print("Key insight: When the disease is rare, even a good test produces many false alarms.")
```

## 6. Independence

### Definition

Events $A$ and $B$ are **independent** if:

$$P(A \cap B) = P(A) \cdot P(B)$$

Equivalently: $P(A|B) = P(A)$ — knowing $B$ doesn't change the probability of $A$.

### Multiplication Rule

For independent events:
$$P(A \cap B) = P(A) \cdot P(B)$$

For dependent events:
$$P(A \cap B) = P(A) \cdot P(B|A)$$

```python
# Example: Independent vs Dependent events

print("Independent Events: Flipping two coins")
print("="*50)
P_head = 0.5
P_both_heads = P_head * P_head
print(f"P(first head) = {P_head}")
print(f"P(second head) = {P_head}")
print(f"P(both heads) = {P_head} × {P_head} = {P_both_heads}")
print()

print("Dependent Events: Drawing cards without replacement")
print("="*50)
P_first_ace = 4/52
P_second_ace_given_first = 3/51
P_both_aces = P_first_ace * P_second_ace_given_first
print(f"P(first ace) = 4/52 = {P_first_ace:.4f}")
print(f"P(second ace | first ace) = 3/51 = {P_second_ace_given_first:.4f}")
print(f"P(both aces) = {P_first_ace:.4f} × {P_second_ace_given_first:.4f} = {P_both_aces:.4f}")
print()
print("Note: After drawing first ace, only 3 remain in 51 cards!")
```

## 7. The Binomial Distribution

### When to Use It

The binomial distribution models the number of successes in $n$ independent trials, each with probability $p$ of success.

### Formula

$$P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}$$

where $\binom{n}{k} = \frac{n!}{k!(n-k)!}$ is the binomial coefficient ("n choose k").

### Parameters

- **Mean:** $\mu = np$
- **Variance:** $\sigma^2 = np(1-p)$
- **Standard deviation:** $\sigma = \sqrt{np(1-p)}$

```python
# Binomial distribution example

print("Example: Survey of 10 people, 30% support a policy")
print("What's the probability exactly 4 people support it?")
print("="*55)
print()

n = 10
p = 0.3
k = 4

# Calculate manually
binom_coeff = comb(n, k)
P_k = binom_coeff * (p**k) * ((1-p)**(n-k))

print(f"P(X = {k}) = C({n},{k}) × {p}^{k} × {1-p}^{n-k}")
print(f"         = {binom_coeff} × {p**k:.6f} × {(1-p)**(n-k):.6f}")
print(f"         = {P_k:.4f}")
print()

# Using scipy
P_scipy = stats.binom.pmf(k, n, p)
print(f"Using scipy: {P_scipy:.4f}")

# Mean and standard deviation
mean = n * p
std = np.sqrt(n * p * (1-p))
print(f"\nExpected value: E[X] = np = {mean:.1f}")
print(f"Standard deviation: σ = √(np(1-p)) = {std:.2f}")
```

```python
# Visualize binomial distribution

n = 10
p_values = [0.3, 0.5, 0.7]
colors = ['blue', 'green', 'red']

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for ax, p, color in zip(axes, p_values, colors):
    k_values = np.arange(0, n+1)
    probabilities = stats.binom.pmf(k_values, n, p)
    
    ax.bar(k_values, probabilities, color=color, alpha=0.7, edgecolor='black')
    ax.axvline(x=n*p, color='red', linestyle='--', linewidth=2, label=f'Mean = {n*p}')
    
    ax.set_xlabel('Number of successes (k)', fontsize=11)
    ax.set_ylabel('Probability P(X = k)', fontsize=11)
    ax.set_title(f'Binomial(n={n}, p={p})', fontsize=12)
    ax.legend()
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("Notice: p = 0.5 gives a symmetric distribution.")
print("p < 0.5 is skewed right, p > 0.5 is skewed left.")
```

## 8. Cumulative Probabilities

### At Most, At Least, Between

- $P(X \leq k)$ — cumulative distribution function (CDF)
- $P(X \geq k) = 1 - P(X < k) = 1 - P(X \leq k-1)$
- $P(a \leq X \leq b) = P(X \leq b) - P(X \leq a-1)$

```python
# Cumulative probability calculations

n = 10
p = 0.3

print("Cumulative Probability Examples")
print(f"X ~ Binomial(n={n}, p={p})")
print("="*45)
print()

# P(X ≤ 3)
P_le_3 = stats.binom.cdf(3, n, p)
print(f"P(X ≤ 3) = {P_le_3:.4f}")

# P(X ≥ 5)
P_ge_5 = 1 - stats.binom.cdf(4, n, p)
print(f"P(X ≥ 5) = 1 - P(X ≤ 4) = {P_ge_5:.4f}")

# P(2 ≤ X ≤ 5)
P_2_to_5 = stats.binom.cdf(5, n, p) - stats.binom.cdf(1, n, p)
print(f"P(2 ≤ X ≤ 5) = P(X ≤ 5) - P(X ≤ 1) = {P_2_to_5:.4f}")

# Visualize
k_values = np.arange(0, n+1)
probs = stats.binom.pmf(k_values, n, p)
cdf = stats.binom.cdf(k_values, n, p)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# PMF
ax1.bar(k_values, probs, color='blue', alpha=0.7, edgecolor='black')
ax1.set_xlabel('k')
ax1.set_ylabel('P(X = k)')
ax1.set_title('Probability Mass Function (PMF)')
ax1.grid(True, alpha=0.3)

# CDF
ax2.step(k_values, cdf, where='mid', color='red', linewidth=2)
ax2.scatter(k_values, cdf, color='red', s=50, zorder=5)
ax2.set_xlabel('k')
ax2.set_ylabel('P(X ≤ k)')
ax2.set_title('Cumulative Distribution Function (CDF)')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
