## 1. The Definite Integral as Area

### Notation

The **definite integral** of $f(x)$ from $a$ to $b$ is written as:

$$\int_a^b f(x) \, dx$$

where:
- $a$ is the **lower limit** of integration
- $b$ is the **upper limit** of integration

### Geometric Interpretation

When $f(x) \geq 0$ on $[a, b]$, the definite integral equals the **area** under the curve $y = f(x)$, above the $x$-axis, between $x = a$ and $x = b$.

```python
# Visualizing the definite integral as area

def f(x):
    return x**2

a, b = 1, 3  # Limits of integration

x = np.linspace(0, 4, 200)
x_fill = np.linspace(a, b, 100)

plt.figure(figsize=(10, 6))
plt.plot(x, f(x), 'b-', linewidth=2, label='f(x) = x²')
plt.fill_between(x_fill, f(x_fill), alpha=0.3, color='blue', label='Area under curve')
plt.axvline(x=a, color='red', linestyle='--', alpha=0.7)
plt.axvline(x=b, color='red', linestyle='--', alpha=0.7)
plt.axhline(y=0, color='gray', linewidth=0.5)

# Labels
plt.text(a, -0.5, 'a = 1', ha='center', fontsize=12)
plt.text(b, -0.5, 'b = 3', ha='center', fontsize=12)

plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.title('Definite Integral = Area Under the Curve', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.ylim(-1, 10)
plt.show()

# Calculate the integral
# ∫ x² dx from 1 to 3 = [x³/3] from 1 to 3 = 27/3 - 1/3 = 26/3
area = (3**3/3) - (1**3/3)
print(f"∫₁³ x² dx = [x³/3]₁³ = (27/3) - (1/3) = {area:.4f}")
```

## 2. The Fundamental Theorem of Calculus

### Part 1: Differentiation of Integrals

If $F(x) = \int_a^x f(t) \, dt$, then $F'(x) = f(x)$

### Part 2: Evaluation of Definite Integrals

If $F$ is any antiderivative of $f$ (so $F'(x) = f(x)$), then:

$$\int_a^b f(x) \, dx = F(b) - F(a) = \left[F(x)\right]_a^b$$

**This is the key result:** To evaluate a definite integral, find an antiderivative and subtract its values at the limits.

```python
# Example: Using the Fundamental Theorem

print("Example: Evaluate ∫₀² (3x² + 2x) dx")
print("="*50)
print()
print("Step 1: Find an antiderivative")
print("  F(x) = x³ + x²")
print()
print("Step 2: Apply the Fundamental Theorem")
print("  ∫₀² (3x² + 2x) dx = [x³ + x²]₀²")
print("                    = F(2) - F(0)")
print("                    = (2³ + 2²) - (0³ + 0²)")
print("                    = (8 + 4) - 0")
print("                    = 12")

# Visualization
x = np.linspace(-0.5, 2.5, 200)
x_fill = np.linspace(0, 2, 100)

def f(x):
    return 3*x**2 + 2*x

plt.figure(figsize=(10, 6))
plt.plot(x, f(x), 'b-', linewidth=2, label='f(x) = 3x² + 2x')
plt.fill_between(x_fill, f(x_fill), alpha=0.3, color='blue', label='Area = 12')
plt.axvline(x=0, color='red', linestyle='--', alpha=0.7)
plt.axvline(x=2, color='red', linestyle='--', alpha=0.7)
plt.axhline(y=0, color='gray', linewidth=0.5)

plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.title('∫₀² (3x² + 2x) dx = 12', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## 3. Properties of Definite Integrals

| Property | Formula |
|----------|--------|
| Reversing limits | $\int_a^b f(x)\,dx = -\int_b^a f(x)\,dx$ |
| Zero interval | $\int_a^a f(x)\,dx = 0$ |
| Constant multiple | $\int_a^b k \cdot f(x)\,dx = k\int_a^b f(x)\,dx$ |
| Sum/difference | $\int_a^b [f(x) \pm g(x)]\,dx = \int_a^b f(x)\,dx \pm \int_a^b g(x)\,dx$ |
| Additivity | $\int_a^b f(x)\,dx + \int_b^c f(x)\,dx = \int_a^c f(x)\,dx$ |

```python
# Demonstrating the additivity property

def f(x):
    return x**2

# ∫₀³ x² dx = ∫₀¹ x² dx + ∫₁³ x² dx

# Using antiderivative F(x) = x³/3
def F(x):
    return x**3 / 3

int_0_3 = F(3) - F(0)
int_0_1 = F(1) - F(0)
int_1_3 = F(3) - F(1)

print("Additivity Property Demonstration")
print("="*45)
print(f"∫₀³ x² dx = {int_0_3:.4f}")
print(f"∫₀¹ x² dx = {int_0_1:.4f}")
print(f"∫₁³ x² dx = {int_1_3:.4f}")
print(f"\n∫₀¹ + ∫₁³ = {int_0_1 + int_1_3:.4f}")
print(f"\nThey match! ✓")

# Visualization
x = np.linspace(0, 3.5, 200)
x_left = np.linspace(0, 1, 50)
x_right = np.linspace(1, 3, 50)

plt.figure(figsize=(10, 6))
plt.plot(x, f(x), 'b-', linewidth=2, label='f(x) = x²')
plt.fill_between(x_left, f(x_left), alpha=0.3, color='green', label=f'∫₀¹ = {int_0_1:.3f}')
plt.fill_between(x_right, f(x_right), alpha=0.3, color='orange', label=f'∫₁³ = {int_1_3:.3f}')
plt.axvline(x=0, color='gray', linestyle='--', alpha=0.5)
plt.axvline(x=1, color='gray', linestyle='--', alpha=0.5)
plt.axvline(x=3, color='gray', linestyle='--', alpha=0.5)
plt.axhline(y=0, color='gray', linewidth=0.5)

plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.title(f'Additivity: ∫₀¹ + ∫₁³ = ∫₀³ = {int_0_3:.3f}', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## 4. Signed Areas and Net Area

### Important Distinction

- When $f(x) < 0$, the integral counts area as **negative**
- The definite integral gives the **net signed area** (above minus below)
- For **total area**, integrate $|f(x)|$ or split the integral at zeros

```python
# Signed area vs total area

def f(x):
    return x**3 - 3*x

# f(x) = 0 when x³ - 3x = 0 → x(x² - 3) = 0 → x = 0, ±√3

x = np.linspace(-2, 2, 200)
x_neg = np.linspace(-np.sqrt(3), 0, 100)
x_pos = np.linspace(0, np.sqrt(3), 100)

# Calculate integrals
# F(x) = x⁴/4 - 3x²/2
def F(x):
    return x**4/4 - 3*x**2/2

int_neg_sqrt3_to_0 = F(0) - F(-np.sqrt(3))
int_0_to_sqrt3 = F(np.sqrt(3)) - F(0)
int_total = F(np.sqrt(3)) - F(-np.sqrt(3))

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Left: Show signed areas
ax1.plot(x, f(x), 'b-', linewidth=2, label='f(x) = x³ - 3x')
ax1.fill_between(x_neg, f(x_neg), alpha=0.3, color='green', label=f'Area above: +{abs(int_neg_sqrt3_to_0):.3f}')
ax1.fill_between(x_pos, f(x_pos), alpha=0.3, color='red', label=f'Area below: {int_0_to_sqrt3:.3f}')
ax1.axhline(y=0, color='gray', linewidth=0.5)
ax1.set_xlabel('x', fontsize=12)
ax1.set_ylabel('f(x)', fontsize=12)
ax1.set_title('Signed Areas', fontsize=14)
ax1.legend()
ax1.grid(True, alpha=0.3)

# Right: Comparison
categories = ['Net Signed Area', 'Total Area']
values = [int_total, abs(int_neg_sqrt3_to_0) + abs(int_0_to_sqrt3)]
colors = ['blue', 'orange']

ax2.bar(categories, values, color=colors, alpha=0.7)
for i, v in enumerate(values):
    ax2.text(i, v + 0.1, f'{v:.3f}', ha='center', fontsize=12)

ax2.set_ylabel('Value', fontsize=12)
ax2.set_title('Net vs Total Area', fontsize=14)

plt.tight_layout()
plt.show()

print(f"Net signed area (∫ f dx): {int_total:.4f}")
print(f"Total area (∫ |f| dx): {abs(int_neg_sqrt3_to_0) + abs(int_0_to_sqrt3):.4f}")
```

## 5. Area Between Two Curves

### Formula

If $f(x) \geq g(x)$ on $[a, b]$, the area between the curves is:

$$A = \int_a^b [f(x) - g(x)] \, dx$$

This is "top curve minus bottom curve."

```python
# Example: Area between two curves

def f(x):
    return -x**2 + 4  # Parabola opening down

def g(x):
    return x          # Line

# Find intersection points: -x² + 4 = x → x² + x - 4 = 0
# x = (-1 ± √17)/2
x1 = (-1 - np.sqrt(17))/2
x2 = (-1 + np.sqrt(17))/2

x = np.linspace(-3, 3, 200)
x_fill = np.linspace(x1, x2, 100)

plt.figure(figsize=(10, 6))
plt.plot(x, f(x), 'b-', linewidth=2, label='f(x) = -x² + 4')
plt.plot(x, g(x), 'r-', linewidth=2, label='g(x) = x')
plt.fill_between(x_fill, f(x_fill), g(x_fill), alpha=0.3, color='purple', label='Area between curves')
plt.scatter([x1, x2], [g(x1), g(x2)], color='green', s=100, zorder=5, label='Intersection points')
plt.axhline(y=0, color='gray', linewidth=0.5)
plt.axvline(x=0, color='gray', linewidth=0.5)

plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('Area Between Two Curves', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.ylim(-5, 5)
plt.show()

# Calculate the area
# ∫[(-x² + 4) - x] dx = ∫(-x² - x + 4) dx = -x³/3 - x²/2 + 4x
def integral(x):
    return -x**3/3 - x**2/2 + 4*x

area = integral(x2) - integral(x1)
print(f"Intersection points: x = {x1:.4f} and x = {x2:.4f}")
print(f"Area = ∫[f(x) - g(x)] dx from {x1:.4f} to {x2:.4f}")
print(f"Area = {area:.4f} square units")
```

## 6. Arithmetic Series

### Definition

An arithmetic sequence has constant difference $d$ between consecutive terms:
$$a_n = a_1 + (n-1)d$$

### Sum Formula

The sum of the first $n$ terms:

$$S_n = \frac{n}{2}(a_1 + a_n) = \frac{n}{2}[2a_1 + (n-1)d]$$

```python
# Example: Arithmetic series

print("Example: Sum of 2 + 5 + 8 + 11 + ... + 50")
print("="*50)
print()
print("Step 1: Identify the sequence")
print("  a₁ = 2, d = 3")
print()
print("Step 2: Find n (number of terms)")
print("  aₙ = a₁ + (n-1)d")
print("  50 = 2 + (n-1)·3")
print("  48 = 3(n-1)")
print("  n = 17")
print()
print("Step 3: Calculate the sum")
print("  Sₙ = n/2 · (a₁ + aₙ)")
print("  S₁₇ = 17/2 · (2 + 50)")
print("      = 8.5 · 52")
print("      = 442")
print()

# Verify by direct calculation
a_1 = 2
d = 3
terms = [a_1 + (n-1)*d for n in range(1, 18)]
print(f"Verification: {' + '.join(map(str, terms[:5]))} + ... + {terms[-1]}")
print(f"Direct sum: {sum(terms)}")
```

## 7. Geometric Series

### Definition

A geometric sequence has constant ratio $r$ between consecutive terms:
$$a_n = a_1 \cdot r^{n-1}$$

### Finite Sum Formula

The sum of the first $n$ terms (for $r \neq 1$):

$$S_n = a_1 \cdot \frac{1 - r^n}{1 - r}$$

### Infinite Sum Formula

For $|r| < 1$, the sum converges:

$$S_\infty = \frac{a_1}{1 - r}$$

```python
# Example: Geometric series

print("Example 1: Sum of 3 + 6 + 12 + 24 + ... (8 terms)")
print("="*50)

a_1 = 3
r = 2
n = 8

print(f"a₁ = {a_1}, r = {r}, n = {n}")
print(f"Sₙ = a₁ · (1 - rⁿ)/(1 - r)")
print(f"S₈ = 3 · (1 - 2⁸)/(1 - 2)")
print(f"   = 3 · (1 - 256)/(-1)")
print(f"   = 3 · 255")
print(f"   = 765")

# Verify
terms = [a_1 * r**(i) for i in range(n)]
print(f"\nVerification: {terms}")
print(f"Sum: {sum(terms)}")

print("\n" + "="*50)
print("Example 2: Infinite sum of 1 + 1/2 + 1/4 + 1/8 + ...")
print("="*50)

a_1 = 1
r = 0.5

print(f"a₁ = {a_1}, r = {r}")
print(f"Since |r| = 0.5 < 1, the series converges.")
print(f"S∞ = a₁/(1 - r) = 1/(1 - 0.5) = 1/0.5 = 2")

# Visualize convergence
n_terms = np.arange(1, 21)
partial_sums = [sum([a_1 * r**(i) for i in range(n)]) for n in n_terms]

plt.figure(figsize=(10, 6))
plt.plot(n_terms, partial_sums, 'bo-', markersize=8, label='Partial sums')
plt.axhline(y=2, color='red', linestyle='--', linewidth=2, label='S∞ = 2')
plt.xlabel('Number of terms', fontsize=12)
plt.ylabel('Partial sum', fontsize=12)
plt.title('Geometric Series Converging to 2', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## 8. Riemann Sums: The Foundation

### The Idea

The definite integral is defined as a limit of **Riemann sums** — approximations using rectangles.

$$\int_a^b f(x) \, dx = \lim_{n \to \infty} \sum_{i=1}^{n} f(x_i^*) \Delta x$$

where $\Delta x = (b-a)/n$

```python
# Visualizing Riemann sums approaching the integral

def f(x):
    return x**2

a, b = 0, 2

# True value: ∫₀² x² dx = [x³/3]₀² = 8/3
true_value = 8/3

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
n_values = [4, 8, 16, 32]

for ax, n in zip(axes.flatten(), n_values):
    dx = (b - a) / n
    x_smooth = np.linspace(a, b, 200)
    
    # Plot the function
    ax.plot(x_smooth, f(x_smooth), 'b-', linewidth=2)
    
    # Draw rectangles (right Riemann sum)
    riemann_sum = 0
    for i in range(n):
        x_left = a + i*dx
        x_right = a + (i+1)*dx
        height = f(x_right)  # Right endpoint
        riemann_sum += height * dx
        rect = plt.Rectangle((x_left, 0), dx, height, 
                            edgecolor='red', facecolor='red', alpha=0.3)
        ax.add_patch(rect)
    
    error = abs(riemann_sum - true_value)
    ax.set_xlabel('x', fontsize=10)
    ax.set_ylabel('f(x)', fontsize=10)
    ax.set_title(f'n = {n} rectangles\nSum = {riemann_sum:.4f}, Error = {error:.4f}', fontsize=11)
    ax.grid(True, alpha=0.3)
    ax.set_xlim(-0.1, 2.1)
    ax.set_ylim(0, 4.5)

plt.suptitle(f'Riemann Sums Approaching ∫₀² x² dx = {true_value:.4f}', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()

print("As n increases, the Riemann sum approaches the true integral.")
print(f"True value: {true_value:.6f}")
```
