## 1. Angles and Radian Measure

### Why Radians?

While degrees are familiar, **radians** are the natural unit for calculus and physics.

### Definition

One **radian** is the angle subtended by an arc equal in length to the radius.

$$2\pi \text{ radians} = 360°$$

### Conversion Formulas

$$\text{radians} = \text{degrees} \times \frac{\pi}{180}$$

$$\text{degrees} = \text{radians} \times \frac{180}{\pi}$$

```python
# Conversion between degrees and radians

def deg_to_rad(deg):
    return deg * np.pi / 180

def rad_to_deg(rad):
    return rad * 180 / np.pi

print("Common Angle Conversions")
print("="*45)
print()
print("Degrees  →  Radians")
print("-"*30)

common_angles = [0, 30, 45, 60, 90, 120, 180, 270, 360]
for deg in common_angles:
    rad = deg_to_rad(deg)
    # Express as fraction of π
    frac = deg / 180
    if frac == 0:
        rad_str = "0"
    elif frac == 1:
        rad_str = "π"
    elif frac == 2:
        rad_str = "2π"
    else:
        # Simplify fraction
        from fractions import Fraction
        f = Fraction(deg, 180).limit_denominator()
        if f.numerator == 1:
            rad_str = f"π/{f.denominator}"
        else:
            rad_str = f"{f.numerator}π/{f.denominator}"
    print(f"{deg:3d}°     →  {rad_str:8s}  ({rad:.4f})")
```

## 2. The Unit Circle

### Definition

The **unit circle** is a circle of radius 1 centered at the origin.

For any angle $\theta$:
- $\cos\theta$ = x-coordinate
- $\sin\theta$ = y-coordinate

### Fundamental Identity

$$\sin^2\theta + \cos^2\theta = 1$$

```python
# Visualize the unit circle

fig, ax = plt.subplots(figsize=(10, 10))

# Draw unit circle
theta = np.linspace(0, 2*np.pi, 100)
ax.plot(np.cos(theta), np.sin(theta), 'b-', linewidth=2)

# Draw axes
ax.axhline(y=0, color='gray', linewidth=0.5)
ax.axvline(x=0, color='gray', linewidth=0.5)

# Mark special angles
special_angles = {
    0: '0',
    np.pi/6: 'π/6',
    np.pi/4: 'π/4',
    np.pi/3: 'π/3',
    np.pi/2: 'π/2',
    2*np.pi/3: '2π/3',
    3*np.pi/4: '3π/4',
    5*np.pi/6: '5π/6',
    np.pi: 'π',
    7*np.pi/6: '7π/6',
    5*np.pi/4: '5π/4',
    4*np.pi/3: '4π/3',
    3*np.pi/2: '3π/2',
    5*np.pi/3: '5π/3',
    7*np.pi/4: '7π/4',
    11*np.pi/6: '11π/6'
}

for angle, label in special_angles.items():
    x, y = np.cos(angle), np.sin(angle)
    ax.scatter([x], [y], color='red', s=50, zorder=5)
    # Position label
    ax.annotate(label, (x*1.15, y*1.15), ha='center', va='center', fontsize=10)

# Example angle at π/4
example_angle = np.pi/4
x_ex, y_ex = np.cos(example_angle), np.sin(example_angle)
ax.plot([0, x_ex], [0, y_ex], 'g-', linewidth=2)
ax.plot([x_ex, x_ex], [0, y_ex], 'g--', linewidth=1, alpha=0.7)
ax.plot([0, x_ex], [y_ex, y_ex], 'g--', linewidth=1, alpha=0.7)
ax.annotate(f'cos(π/4) = {x_ex:.3f}', (x_ex/2, -0.1), fontsize=11, color='green')
ax.annotate(f'sin(π/4) = {y_ex:.3f}', (x_ex+0.1, y_ex/2), fontsize=11, color='green')

ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_aspect('equal')
ax.set_title('The Unit Circle', fontsize=14)
ax.grid(True, alpha=0.3)
plt.show()

# Print key values
print("Key Values on the Unit Circle:")
print("="*45)
print("θ        cos(θ)    sin(θ)")
print("-"*35)
for angle_rad, angle_str in [(0, '0'), (np.pi/6, 'π/6'), (np.pi/4, 'π/4'), 
                              (np.pi/3, 'π/3'), (np.pi/2, 'π/2')]:
    print(f"{angle_str:8s} {np.cos(angle_rad):8.4f} {np.sin(angle_rad):8.4f}")
```

## 3. Graphs of Sine and Cosine

### Basic Properties

| Property | sin(x) | cos(x) |
|----------|--------|--------|
| Domain | All real numbers | All real numbers |
| Range | [-1, 1] | [-1, 1] |
| Period | 2π | 2π |
| Amplitude | 1 | 1 |
| y-intercept | 0 | 1 |

```python
# Graphs of sin and cos

x = np.linspace(-2*np.pi, 2*np.pi, 500)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))

# Sine
ax1.plot(x, np.sin(x), 'b-', linewidth=2, label='y = sin(x)')
ax1.axhline(y=0, color='gray', linewidth=0.5)
ax1.axhline(y=1, color='gray', linestyle='--', linewidth=0.5, alpha=0.7)
ax1.axhline(y=-1, color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

# Mark key points
key_x = [0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi]
key_y = [np.sin(xi) for xi in key_x]
ax1.scatter(key_x, key_y, color='red', s=80, zorder=5)

ax1.set_xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi])
ax1.set_xticklabels(['-2π', '-π', '0', 'π', '2π'])
ax1.set_ylabel('y', fontsize=12)
ax1.set_title('Graph of y = sin(x)', fontsize=14)
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.set_ylim(-1.5, 1.5)

# Cosine
ax2.plot(x, np.cos(x), 'r-', linewidth=2, label='y = cos(x)')
ax2.axhline(y=0, color='gray', linewidth=0.5)
ax2.axhline(y=1, color='gray', linestyle='--', linewidth=0.5, alpha=0.7)
ax2.axhline(y=-1, color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

# Mark key points
key_y_cos = [np.cos(xi) for xi in key_x]
ax2.scatter(key_x, key_y_cos, color='blue', s=80, zorder=5)

ax2.set_xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi])
ax2.set_xticklabels(['-2π', '-π', '0', 'π', '2π'])
ax2.set_xlabel('x', fontsize=12)
ax2.set_ylabel('y', fontsize=12)
ax2.set_title('Graph of y = cos(x)', fontsize=14)
ax2.legend()
ax2.grid(True, alpha=0.3)
ax2.set_ylim(-1.5, 1.5)

plt.tight_layout()
plt.show()

print("Key observation: cos(x) = sin(x + π/2)")
print("The cosine is a phase-shifted sine!")
```

## 4. General Sinusoidal Functions

### The General Form

$$y = A\sin(B(x - C)) + D$$

or equivalently

$$y = A\cos(B(x - C)) + D$$

### Parameters

| Parameter | Effect | Formula |
|-----------|--------|--------|
| A | **Amplitude** (vertical stretch) | Amplitude = |A| |
| B | **Frequency** (horizontal compression) | Period = 2π/|B| |
| C | **Phase Shift** (horizontal translation) | Shift C to the right |
| D | **Vertical Shift** (midline) | Midline at y = D |

```python
# Exploring the effect of each parameter

x = np.linspace(0, 4*np.pi, 500)

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Top-left: Varying amplitude (A)
ax = axes[0, 0]
for A in [0.5, 1, 2]:
    ax.plot(x, A * np.sin(x), linewidth=2, label=f'A = {A}')
ax.axhline(y=0, color='gray', linewidth=0.5)
ax.set_title('Effect of Amplitude (A)', fontsize=13)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
ax.grid(True, alpha=0.3)

# Top-right: Varying period (B)
ax = axes[0, 1]
for B in [0.5, 1, 2]:
    period = 2*np.pi / B
    ax.plot(x, np.sin(B * x), linewidth=2, label=f'B = {B} (Period = {period/np.pi:.1f}π)')
ax.axhline(y=0, color='gray', linewidth=0.5)
ax.set_title('Effect of Frequency (B)', fontsize=13)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
ax.grid(True, alpha=0.3)

# Bottom-left: Varying phase shift (C)
ax = axes[1, 0]
for C in [0, np.pi/4, np.pi/2]:
    ax.plot(x, np.sin(x - C), linewidth=2, label=f'C = {C/np.pi:.2f}π')
ax.axhline(y=0, color='gray', linewidth=0.5)
ax.set_title('Effect of Phase Shift (C)', fontsize=13)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
ax.grid(True, alpha=0.3)

# Bottom-right: Varying vertical shift (D)
ax = axes[1, 1]
for D in [-1, 0, 1]:
    ax.plot(x, np.sin(x) + D, linewidth=2, label=f'D = {D}')
ax.axhline(y=0, color='gray', linewidth=0.5)
ax.set_title('Effect of Vertical Shift (D)', fontsize=13)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 5. Determining Parameters from a Graph

### Steps to Find Parameters

1. **Amplitude A:** Half the distance from max to min: $A = \frac{\text{max} - \text{min}}{2}$
2. **Vertical Shift D:** Midline value: $D = \frac{\text{max} + \text{min}}{2}$
3. **Period T:** Distance between consecutive peaks, then $B = \frac{2\pi}{T}$
4. **Phase Shift C:** Where the "standard" pattern starts

```python
# Example: Finding parameters from data

print("Example: Finding Parameters")
print("="*50)
print()
print("Given data:")
print("  Maximum value: y_max = 5")
print("  Minimum value: y_min = 1")
print("  Period: T = π")
print("  Maximum occurs at x = π/4")
print()

y_max = 5
y_min = 1
T = np.pi
x_max = np.pi/4

# Calculate parameters
A = (y_max - y_min) / 2
D = (y_max + y_min) / 2
B = 2 * np.pi / T

# For cosine, maximum occurs at C (no shift needed if using cosine)
# y = A*cos(B*(x - C)) + D, max when B*(x-C) = 0, i.e., x = C
C = x_max

print("Calculated parameters:")
print(f"  Amplitude A = (5 - 1)/2 = {A}")
print(f"  Vertical shift D = (5 + 1)/2 = {D}")
print(f"  B = 2π/T = 2π/π = {B}")
print(f"  Phase shift C = {C/np.pi:.2f}π (using cosine)")
print()
print(f"Final equation: y = {A}cos({B:.0f}(x - π/4)) + {D}")

# Verify with plot
x = np.linspace(0, 2*np.pi, 200)
y = A * np.cos(B * (x - C)) + D

plt.figure(figsize=(10, 5))
plt.plot(x, y, 'b-', linewidth=2, label=f'y = {A}cos({B:.0f}(x - π/4)) + {D}')
plt.axhline(y=D, color='gray', linestyle='--', alpha=0.7, label=f'Midline y = {D}')
plt.axhline(y=y_max, color='red', linestyle=':', alpha=0.7)
plt.axhline(y=y_min, color='red', linestyle=':', alpha=0.7)
plt.scatter([x_max], [y_max], color='green', s=100, zorder=5, label='Maximum')

plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('Constructed Sinusoidal Function', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## 6. Applications: Modeling Periodic Phenomena

Sinusoidal functions model many natural cycles:
- Daily temperature variation
- Seasonal daylight hours
- Tides
- Sound and light waves
- Predator-prey populations

```python
# Example: Modeling daylight hours

print("Application: Daylight Hours in Sydney")
print("="*50)
print()

# Model: D(t) = A*cos(B*(t - C)) + D
# where t = day of year (0 = Jan 1)

# Data for Sydney:
max_daylight = 14.4  # hours (December solstice)
min_daylight = 9.9   # hours (June solstice)

# Calculate parameters
A = (max_daylight - min_daylight) / 2
D = (max_daylight + min_daylight) / 2
T = 365  # Period = 1 year
B = 2 * np.pi / T

# In Southern Hemisphere, max daylight is in late December (day ~355)
# cos reaches max at argument 0, so we need: B*(355 - C) = 0
C = 355  # Approximately Dec 21

print(f"Parameters:")
print(f"  A = {A:.2f} hours (amplitude)")
print(f"  D = {D:.2f} hours (average daylight)")
print(f"  Period = {T} days")
print()

def daylight(t):
    return A * np.cos(B * (t - C)) + D

# Plot
days = np.arange(0, 366)
hours = daylight(days)

# Month labels
month_starts = [0, 31, 59, 90, 120, 151, 181, 212, 243, 273, 304, 334]
month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

plt.figure(figsize=(12, 6))
plt.plot(days, hours, 'b-', linewidth=2)
plt.axhline(y=D, color='gray', linestyle='--', alpha=0.7, label=f'Average = {D:.1f} hrs')

# Mark solstices
june_solstice = 172  # ~June 21
dec_solstice = 355   # ~Dec 21
plt.scatter([june_solstice], [daylight(june_solstice)], color='blue', s=100, zorder=5, label='Winter Solstice')
plt.scatter([dec_solstice], [daylight(dec_solstice)], color='red', s=100, zorder=5, label='Summer Solstice')

plt.xticks(month_starts, month_names)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Hours of Daylight', fontsize=12)
plt.title('Daylight Hours in Sydney Throughout the Year', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.ylim(9, 15)
plt.show()

# Calculate daylight for specific dates
print("\nDaylight predictions:")
dates = [(1, 'Jan 1'), (80, 'Mar 21 (Equinox)'), (172, 'Jun 21 (Winter Solstice)'), 
         (266, 'Sep 23 (Equinox)'), (355, 'Dec 21 (Summer Solstice)')]
for day, name in dates:
    print(f"  {name}: {daylight(day):.1f} hours")
```

## 7. Trigonometric Identities

### Fundamental Identities

| Identity | Formula |
|----------|--------|
| Pythagorean | $\sin^2\theta + \cos^2\theta = 1$ |
| Tangent | $\tan\theta = \frac{\sin\theta}{\cos\theta}$ |
| Reciprocals | $\csc\theta = \frac{1}{\sin\theta}$, $\sec\theta = \frac{1}{\cos\theta}$, $\cot\theta = \frac{1}{\tan\theta}$ |
| Even/Odd | $\cos(-\theta) = \cos\theta$, $\sin(-\theta) = -\sin\theta$ |
| Cofunction | $\sin\theta = \cos(\frac{\pi}{2} - \theta)$ |

```python
# Verifying identities numerically

print("Verifying Trigonometric Identities")
print("="*50)
print()

theta = np.pi/3  # Test angle: 60°
print(f"Test angle: θ = π/3 = {np.degrees(theta):.0f}°")
print()

# Pythagorean identity
lhs = np.sin(theta)**2 + np.cos(theta)**2
print(f"1. Pythagorean: sin²θ + cos²θ = {lhs:.10f} = 1 ✓")

# Tangent definition
tan_def = np.sin(theta) / np.cos(theta)
tan_direct = np.tan(theta)
print(f"2. Tangent: sin(θ)/cos(θ) = {tan_def:.6f}")
print(f"            tan(θ) = {tan_direct:.6f} ✓")

# Even/odd
cos_neg = np.cos(-theta)
cos_pos = np.cos(theta)
print(f"3. Even: cos(-θ) = {cos_neg:.6f}")
print(f"         cos(θ) = {cos_pos:.6f} ✓")

# Cofunction
sin_theta = np.sin(theta)
cos_complement = np.cos(np.pi/2 - theta)
print(f"4. Cofunction: sin(θ) = {sin_theta:.6f}")
print(f"               cos(π/2 - θ) = {cos_complement:.6f} ✓")
```

## 8. Solving Basic Trigonometric Equations

### Strategy

To solve $\sin(x) = k$ or $\cos(x) = k$:
1. Find the reference angle using inverse trig
2. Determine all angles in $[0, 2\pi)$ with that sine/cosine
3. Add $2\pi n$ for all solutions

```python
# Solving sin(x) = 0.5

print("Solving: sin(x) = 0.5")
print("="*45)
print()

k = 0.5

# Reference angle
ref_angle = np.arcsin(k)
print(f"Step 1: Reference angle = arcsin(0.5) = {ref_angle:.4f} = π/6")
print()

# Solutions in [0, 2π)
x1 = ref_angle
x2 = np.pi - ref_angle
print(f"Step 2: Solutions in [0, 2π):")
print(f"  x₁ = π/6 = {x1:.4f}")
print(f"  x₂ = π - π/6 = 5π/6 = {x2:.4f}")
print()

print("Step 3: General solution:")
print(f"  x = π/6 + 2πn  or  x = 5π/6 + 2πn, for integer n")

# Visualize
x = np.linspace(0, 4*np.pi, 500)

plt.figure(figsize=(12, 5))
plt.plot(x, np.sin(x), 'b-', linewidth=2, label='y = sin(x)')
plt.axhline(y=0.5, color='red', linestyle='--', linewidth=2, label='y = 0.5')

# Mark solutions
solutions = [np.pi/6, 5*np.pi/6, np.pi/6 + 2*np.pi, 5*np.pi/6 + 2*np.pi]
plt.scatter(solutions, [0.5]*4, color='green', s=100, zorder=5, label='Solutions')

plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('Solving sin(x) = 0.5', fontsize=14)
plt.xticks([0, np.pi/6, 5*np.pi/6, np.pi, 2*np.pi, 13*np.pi/6, 17*np.pi/6, 3*np.pi, 4*np.pi],
           ['0', 'π/6', '5π/6', 'π', '2π', '13π/6', '17π/6', '3π', '4π'])
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
