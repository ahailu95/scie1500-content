## 1. Real Numbers and Number Sets

Scientists work with different types of numbers, and understanding their relationships helps avoid errors.

### The Number Set Hierarchy

$$\mathbb{N} \subset \mathbb{Z} \subset \mathbb{Q} \subset \mathbb{R}$$

**Number Sets:**

- **$\mathbb{N}$ (Natural numbers):** Counting numbers — 1, 2, 3, 4, ...
- **$\mathbb{Z}$ (Integers):** Whole numbers (positive, negative, zero) — ..., -2, -1, 0, 1, 2, ...
- **$\mathbb{Q}$ (Rational numbers):** Fractions of integers — $\frac{1}{2}$, $-\frac{3}{4}$, 0.75, $2.\overline{3}$
- **$\mathbb{R}$ (Real numbers):** All points on the number line — $\pi$, $\sqrt{2}$, $e$, -4.7

**Key insight:** Every natural number is an integer, every integer is rational, and every rational number is real. But not every real number is rational—numbers like $\pi$ and $\sqrt{2}$ are **irrational** (cannot be expressed as a fraction).

### Set Notation

Scientists often need to describe collections of numbers:

- $\{1, 2, 3\}$ — A set containing exactly these elements
- $\{x : x > 0\}$ — "The set of all x such that x is greater than 0"
- $\{x \in \mathbb{Z} : x > 0\}$ — "All integers x such that x is greater than 0" = $\{1, 2, 3, 4, ...\}$
- $\emptyset$ or $\{\}$ — The empty set

```python
# Example: Working with different number types
natural = 5
integer = -3
rational = 3/4
irrational = np.pi

print(f"Natural number: {natural}")
print(f"Integer: {integer}")
print(f"Rational number: {rational}")
print(f"Irrational number (π): {irrational:.10f}")
print(f"\nπ is irrational - its decimal expansion never repeats or ends!")

# --- Building on the flow: Type interactions ---

# Multiplying a rational by a natural results in a float (real number)
result = rational * natural 
print(f"\nCalculation: {rational} * {natural} = {result}")
print(f"The result type is: {type(result).__name__}")

# Checking if π can be 'contained' by an integer
print(f"Truncated π for integer operations: {int(irrational)}")
```

## 2. Order of Operations (BODMAS/PEMDAS)

When an expression contains multiple operations, the order matters.

### BODMAS Rule

Operations are evaluated in this order (highest to lowest priority):

1. **B** — Brackets (parentheses)
2. **O** — Orders (exponents, roots)
3. **D** — Division
4. **M** — Multiplication
5. **A** — Addition
6. **S** — Subtraction

**Note:** Division and Multiplication have equal priority (evaluate left to right). Same for Addition and Subtraction.

### Examples

**Example 1:** $3 + 4 \times 2$
- Multiplication first: $4 \times 2 = 8$
- Then addition: $3 + 8 = 11$
- **Answer:** 11 (not 14)

**Example 2:** $(3 + 4) \times 2$
- Brackets first: $3 + 4 = 7$
- Then multiplication: $7 \times 2 = 14$
- **Answer:** 14

**Example 3:** $2^3 + 4 \times 5 - 10 \div 2$
- Exponent: $2^3 = 8$
- Multiplication: $4 \times 5 = 20$
- Division: $10 \div 2 = 5$
- Left to right: $8 + 20 - 5 = 23$
- **Answer:** 23

```python
# Python Follows BODMAS
print("BODMAS Examples:")
print(f"3 + 4 * 2 = {3 + 4 * 2}")        # Output: 11
print(f"(3 + 4) * 2 = {(3 + 4) * 2}")    # Output: 14
print(f"2**3 + 4*5 - 10/2 = {2**3 + 4*5 - 10/2}")  # Output: 23.0
```

### ⚖️ Logic: Order of Operations (BODMAS/PEMDAS)
When writing complex formulas in Python, you must follow the correct mathematical order of operations to ensure accuracy.

## 3. Exponents and Radicals

Exponents (powers) and radicals (roots) appear throughout scientific models.

### Exponent Rules

- **Product rule:** $a^m \cdot a^n = a^{m+n}$ — Example: $2^3 \cdot 2^4 = 2^7 = 128$
- **Quotient rule:** $\frac{a^m}{a^n} = a^{m-n}$ — Example: $\frac{5^6}{5^2} = 5^4 = 625$
- **Power rule:** $(a^m)^n = a^{mn}$ — Example: $(3^2)^4 = 3^8 = 6561$
- **Zero exponent:** $a^0 = 1$ (for $a \neq 0$) — Example: $7^0 = 1$
- **Negative exponent:** $a^{-n} = \frac{1}{a^n}$ — Example: $2^{-3} = \frac{1}{8}$
- **Fractional exponent:** $a^{1/n} = \sqrt[n]{a}$ — Example: $8^{1/3} = \sqrt[3]{8} = 2$
- **General fractional:** $a^{m/n} = \sqrt[n]{a^m}$ — Example: $4^{3/2} = \sqrt{4^3} = 8$

### Radical Rules

- **Product:** $\sqrt{ab} = \sqrt{a} \cdot \sqrt{b}$ — Example: $\sqrt{12} = \sqrt{4 \cdot 3} = 2\sqrt{3}$
- **Quotient:** $\sqrt{\frac{a}{b}} = \frac{\sqrt{a}}{\sqrt{b}}$ — Example: $\sqrt{\frac{9}{16}} = \frac{3}{4}$
- **Power:** $\sqrt{a^2} = |a|$ — Example: $\sqrt{(-3)^2} = 3$

```python
# Python Syntax for Exponents

# Exponentiation
print("Exponentiation:")
print(f"2 ** 3 = {2 ** 3}")             # 8 (2 to the power 3)
print(f"math.pow(2, 3) = {math.pow(2, 3)}")  # 8.0 (returns float)

# Square root
print("\nSquare root:")
print(f"math.sqrt(16) = {math.sqrt(16)}")    # 4.0
print(f"16 ** 0.5 = {16 ** 0.5}")            # 4.0 (alternative)

# Cube root
print(f"\n27 ** (1/3) = {27 ** (1/3)}")      # 3.0

# Negative exponent
print(f"\n2 ** -3 = {2 ** -3}")              # 0.125 (= 1/8)
```

## 4. Algebraic and Rational Expressions

### Algebraic Expressions

An algebraic expression combines numbers, variables, and operations.

**Simplifying by combining like terms:**

$$3x^2 + 5x - 2x^2 + 7 - 3x = (3-2)x^2 + (5-3)x + 7 = x^2 + 2x + 7$$

**Expanding brackets:**

$$(a + b)(c + d) = ac + ad + bc + bd$$

**Example:** $(2x + 3)(x - 4) = 2x^2 - 8x + 3x - 12 = 2x^2 - 5x - 12$

### Factoring Common Patterns

- **Difference of squares:** $a^2 - b^2 = (a+b)(a-b)$
- **Perfect square:** $a^2 + 2ab + b^2 = (a+b)^2$
- **Perfect square:** $a^2 - 2ab + b^2 = (a-b)^2$

**Example:** Factor $x^2 - 9$
- Recognize: $x^2 - 9 = x^2 - 3^2$
- Apply difference of squares: $(x+3)(x-3)$

### Rational Expressions

A rational expression is a fraction with polynomials in the numerator and/or denominator.

**Simplifying:** Factor and cancel common terms.

$$\frac{x^2 - 4}{x + 2} = \frac{(x+2)(x-2)}{x+2} = x - 2 \quad \text{(for } x \neq -2\text{)}$$

**Important:** Always note restrictions! The original expression is undefined at $x = -2$.

```python
# Verifying algebraic identities with Python
x = 5  # Test value

# Difference of squares: x² - 9 = (x+3)(x-3)
left_side = x**2 - 9
right_side = (x + 3) * (x - 3)
print(f"Difference of squares (x={x}):")
print(f"  x² - 9 = {left_side}")
print(f"  (x+3)(x-3) = {right_side}")
print(f"  Equal? {left_side == right_side}")

# Rational expression: (x²-4)/(x+2) = x-2
numerator = x**2 - 4
denominator = x + 2
simplified = x - 2
print(f"\nRational expression (x={x}):")
print(f"  (x²-4)/(x+2) = {numerator}/{denominator} = {numerator/denominator}")
print(f"  x - 2 = {simplified}")
```

## 5. Equations

An equation states that two expressions are equal. Solving means finding the value(s) that make it true.

### Linear Equations

Form: $ax + b = c$

**Strategy:** Isolate the variable.

**Example:** Solve $3x + 7 = 22$
1. Subtract 7: $3x = 15$
2. Divide by 3: $x = 5$

### Quadratic Equations

Form: $ax^2 + bx + c = 0$

**Methods:**
1. **Factoring** (when possible)
2. **Quadratic formula:** $x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$

**Example:** Solve $x^2 - 5x + 6 = 0$
- Factor: $(x-2)(x-3) = 0$
- Solutions: $x = 2$ or $x = 3$

**Example:** Solve $2x^2 + 3x - 5 = 0$
- Use quadratic formula with $a=2$, $b=3$, $c=-5$
- $x = \frac{-3 \pm \sqrt{9 + 40}}{4} = \frac{-3 \pm 7}{4}$
- Solutions: $x = 1$ or $x = -2.5$

### The Discriminant

The expression $b^2 - 4ac$ (under the square root) determines the nature of solutions:

- $b^2 - 4ac > 0$ → Two distinct real solutions
- $b^2 - 4ac = 0$ → One repeated real solution
- $b^2 - 4ac < 0$ → No real solutions (complex only)

```python
# Quadratic formula implementation
def solve_quadratic(a, b, c):
    """Solve ax² + bx + c = 0 using the quadratic formula."""
    discriminant = b**2 - 4*a*c
    
    if discriminant > 0:
        x1 = (-b + math.sqrt(discriminant)) / (2*a)
        x2 = (-b - math.sqrt(discriminant)) / (2*a)
        return x1, x2, "Two distinct real solutions"
    elif discriminant == 0:
        x = -b / (2*a)
        return x, x, "One repeated real solution"
    else:
        return None, None, "No real solutions (complex only)"

# Example 1: x² - 5x + 6 = 0
x1, x2, msg = solve_quadratic(1, -5, 6)
print(f"x² - 5x + 6 = 0: x = {x1}, {x2}")
print(f"  {msg}\n")

# Example 2: 2x² + 3x - 5 = 0
x1, x2, msg = solve_quadratic(2, 3, -5)
print(f"2x² + 3x - 5 = 0: x = {x1}, {x2}")
print(f"  {msg}")
```

## 6. Inequalities

Inequalities compare expressions using $<$, $>$, $\leq$, $\geq$.

### Solving Linear Inequalities

**Rules:**
- Add/subtract the same quantity to both sides: inequality preserved
- Multiply/divide by a **positive** number: inequality preserved
- Multiply/divide by a **negative** number: **inequality reverses**

**Example:** Solve $-2x + 5 > 11$
1. Subtract 5: $-2x > 6$
2. Divide by $-2$ (reverse inequality!): $x < -3$

**Solution:** $x < -3$ or $(-\infty, -3)$

### Compound Inequalities

**Example:** Solve $-3 \leq 2x + 1 < 7$
1. Subtract 1 from all parts: $-4 \leq 2x < 6$
2. Divide by 2: $-2 \leq x < 3$

**Solution:** $-2 \leq x < 3$ or $[-2, 3)$

```python
# Visualizing inequality solution: -2x + 5 > 11
# Solution: x < -3

x = np.linspace(-8, 2, 100)
y = -2*x + 5

plt.figure(figsize=(10, 5))
plt.plot(x, y, 'b-', linewidth=2, label='y = -2x + 5')
plt.axhline(y=11, color='r', linestyle='--', 
            linewidth=2, label='y = 11')
plt.fill_between(x, y, 11, where=(y > 11), 
                 alpha=0.3, color='green', 
                 label='Solution region (x < -3)')
plt.axvline(x=-3, color='g', linestyle=':', linewidth=2)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('Inequality: -2x + 5 > 11 → x < -3', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## 7. Scientific Notation and Unit Conversions

### Scientific Notation

Large and small numbers are written as: $a \times 10^n$ where $1 \leq |a| < 10$

**Examples:**
- 4,500,000 → $4.5 \times 10^6$
- 0.00032 → $3.2 \times 10^{-4}$
- 381,000,000 (tonnes) → $3.81 \times 10^8$ tonnes

**Operations:**
- Multiply: $(2 \times 10^3)(3 \times 10^4) = 6 \times 10^7$
- Divide: $\frac{8 \times 10^5}{2 \times 10^2} = 4 \times 10^3$

### Unit Conversions

Scientists constantly convert between units.

**Example:** Convert 381,000,000 metric tonnes to million metric tonnes (MMT)

$$381,000,000 \text{ MT} = \frac{381,000,000}{1,000,000} \text{ MMT} = 381 \text{ MMT}$$

**Example:** Convert 5.1 kg/person/day to grams/person/day

$$5.1 \text{ kg/person/day} \times \frac{1000 \text{ g}}{1 \text{ kg}} = 5100 \text{ g/person/day}$$

```python
# Scientific notation in Python
large = 3.81e8    # 3.81 × 10^8
small = 3.2e-4    # 3.2 × 10^-4

print("Scientific Notation:")
print(f"3.81e8 = {large}")       # 381000000.0
print(f"3.2e-4 = {small}")       # 0.00032

# Unit conversion
tonnes = 381_000_000  # Using underscore for readability
mmt = tonnes / 1_000_000
print(f"\n{tonnes:,} tonnes = {mmt} MMT")

# kg to grams
kg_per_day = 5.1
g_per_day = kg_per_day * 1000
print(f"{kg_per_day} kg/person/day = {g_per_day} g/person/day")
```

## 8. Function Notation: What $f(x)$ Actually Means

This is crucial preparation for Week 1 and beyond.

### The Meaning of $f(x)$

When we write $f(x) = 2x + 3$, we're defining a **rule** called $f$ that takes an input $x$ and produces an output.

**$f(x)$ does NOT mean "$f$ times $x$"!**

It means "the function $f$ evaluated at the value $x$."

### Evaluating Functions

If $f(x) = 2x + 3$:
- $f(5) = 2(5) + 3 = 13$ (input 5, output 13)
- $f(-2) = 2(-2) + 3 = -1$ (input -2, output -1)
- $f(a) = 2a + 3$ (input $a$, output $2a + 3$)
- $f(x+1) = 2(x+1) + 3 = 2x + 5$ (input $x+1$, output $2x+5$)

### Multiple Functions

We can have several functions with different names:
- $f(x) = x^2$
- $g(x) = 3x - 1$
- $h(t) = 5t + 2$ (using $t$ as the variable)

Then:
- $f(3) = 9$
- $g(3) = 8$
- $h(3) = 17$

```python
# Python Functions Mirror Mathematical Functions

# Define a function
def f(x):
    return 2*x + 3

# Evaluate it
print("f(x) = 2x + 3")
print(f"f(5) = {f(5)}")     # 13
print(f"f(-2) = {f(-2)}")   # -1
print(f"f(0) = {f(0)}")     # 3

# Multiple functions
def g(x):
    return 3*x - 1

def h(t):
    return 5*t + 2

print(f"\ng(3) = {g(3)}")   # 8
print(f"h(3) = {h(3)}")     # 17

# This direct parallel between mathematical notation 
# and Python code is why learning Python helps 
# reinforce mathematical understanding.
```

## 9. Interval Notation

In Week 1, you'll describe **domains** (valid inputs) and **ranges** (possible outputs) of functions. Interval notation is the standard way to do this.

### Types of Intervals

- **Closed** $[a, b]$ means $a \leq x \leq b$ — both endpoints included (●────●)
- **Open** $(a, b)$ means $a < x < b$ — both endpoints excluded (○────○)
- **Half-open (left)** $(a, b]$ means $a < x \leq b$ (○────●)
- **Half-open (right)** $[a, b)$ means $a \leq x < b$ (●────○)
- **Infinite** $[a, \infty)$ means $x \geq a$ (●────→)
- **Infinite** $(-\infty, b)$ means $x < b$ (←────○)
- **All reals** $(-\infty, \infty)$ means all $x \in \mathbb{R}$ (←────→)

**Key points:**
- Square bracket $[$ or $]$ means **included** (≤ or ≥)
- Round bracket $($ or $)$ means **excluded** (< or >)
- Infinity ($\infty$) always uses round brackets (you can't "reach" infinity)

### Examples

- "All numbers from 0 to 10, including both endpoints" → $[0, 10]$
- "All positive numbers" → $(0, \infty)$
- "All numbers less than or equal to 5" → $(-\infty, 5]$
- "All numbers except 3" → $(-\infty, 3) \cup (3, \infty)$

```python
# Visualizing interval notation
fig, axes = plt.subplots(4, 1, figsize=(10, 6))

intervals = [
    ("[2, 5] (Closed)", 2, 5, 'both'),
    ("(2, 5) (Open)", 2, 5, 'neither'),
    ("[2, 5) (Half-open)", 2, 5, 'left'),
    ("(2, ∞) (Infinite)", 2, 8, 'left_open'),
]

for ax, (label, a, b, style) in zip(axes, intervals):
    ax.set_xlim(0, 10)
    ax.set_ylim(-0.5, 0.5)
    ax.axhline(y=0, color='gray', linewidth=1)
    
    # Draw the interval line
    ax.plot([a, b], [0, 0], 'b-', linewidth=4)
    
    # Draw endpoints
    if style == 'both':
        ax.plot(a, 0, 'bo', markersize=12)
        ax.plot(b, 0, 'bo', markersize=12)
    elif style == 'neither':
        ax.plot(a, 0, 'bo', markersize=12, 
                markerfacecolor='white')
        ax.plot(b, 0, 'bo', markersize=12, 
                markerfacecolor='white')
    elif style == 'left':
        ax.plot(a, 0, 'bo', markersize=12)
        ax.plot(b, 0, 'bo', markersize=12, 
                markerfacecolor='white')
    elif style == 'left_open':
        ax.plot(a, 0, 'bo', markersize=12, 
                markerfacecolor='white')
        ax.annotate('', xy=(10, 0), xytext=(b, 0),
                    arrowprops=dict(arrowstyle='->', 
                                    color='blue', lw=4))
    
    ax.set_ylabel(label, fontsize=11)
    ax.set_yticks([])
    ax.set_xticks(range(11))

plt.tight_layout()
plt.suptitle('Interval Notation Visualized', 
             fontsize=14, y=1.02)
plt.show()
```

## 10. Why This Matters: Equations in Scientific Modelling

Everything you've reviewed connects to real scientific work. Here's a preview of how equations model the natural world.

### The Temperature Conversion Model

The relationship between Fahrenheit and Celsius temperatures:

$$C = \frac{5}{9}(F - 32)$$

This is a **linear equation**. Given any Fahrenheit value, we can compute the Celsius equivalent.

### The Ocean Plastic Crisis

Scientists estimate that **4.8 to 12.7 million metric tonnes** of plastic entered the oceans in 2010 alone (Jambeck et al., 2015). To model this, they use:

- **Linear equations** to describe simple relationships
- **Quadratic equations** to model growth rates (like fish populations)
- **Exponential equations** to model accelerating growth (like plastic production)

The production of plastic has grown from 2 million tonnes in 1950 to 381 million tonnes in 2015—an increase of over 190×. This is clearly not linear growth.

### Coming Up: Bounded Growth Models

In **Week 3**, you'll encounter the **Schaefer fish growth model** — an elegant equation that uses quadratics to describe how fish populations grow when limited by their environment. The mathematical tools you're building this week (algebra, quadratics, BODMAS) are exactly what you'll need.


```python
# Temperature Conversion Model
def fahrenheit_to_celsius(F):
    return (5/9) * (F - 32)

temps_f = [32, 50, 68, 86, 100, 212]
print("Temperature Conversion:")
for f in temps_f:
    c = fahrenheit_to_celsius(f)
    print(f"  {f}°F = {c:.1f}°C")
```
