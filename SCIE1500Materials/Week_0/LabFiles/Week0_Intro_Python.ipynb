### 2. Learn the Python Syntax

Now let's start learning how to write Python instructions.

Human languages work because they follow a set of rules or a `syntax` that controls how words, symbols and punctation marks are arranged to generate meaning. Computer languages are the same--they have syntax. If the syntax is not followed, you get syntax error: your instructions cannot be interpreted and the computer cannot execute them. The difference between human and computer languages is that the latter are unforgiving and there is no margin for error ('computers cannot read typos'). Computer languages use a very `specific set of words`. Further, most major programming languages (e.g. C/C++, Java, and Python) are `case sensitive`. For example, 'print' is not the same as 'Print', 'Input' the same as 'input', or 'Type' the same as 'type'. Therefore, it is essential that you pay attention to detail as you learn about the language. 

#### 2.1 Python functions are like functions in algebra

Even when you know what specific words/commands to use, you need to know how to use them. For example, to print `Ftemps` above, we wrote: `print(Ftemps)`. Why did we not write `print Ftemps`? 


The answer is that the `print` function is like the function in algebra (`f(x)`) that takes a value passed to it inside parentheses (e.g. `x`). Therefore, always remember to pass arguments to a Python function inside parentheses.

Examples:

```python
#print 2 + 2
print(2+2)
#print 3x2
print(3*2)
#print nothing
print() #we still need parenthesis
#print 3 squared
print(3**2)
```

The following code is not syntactically correct (will generate errors if run). The problem is that there are no parentheses around the arguments being passed to the `print` command. 

```python
print 2+2
print 3*2
print 3**2
```

#### 2.3 Comments

Comments are an essential part of a computer program -- they help document/explain the instructions by providing a running commentary. Comment lines are distinguished from actual Python instruction lines by a hash mark/tag. Any thing that follows a hash tag is ignored by Python (it is stripped out and not considered as instruction).

```python
# I am a comment line
# and the following instruction is commented out
# print(2*5)

#But the line below will be exectuded
print(3*17)

# now we are done demonstrating a hash tag
```

#### 2.4 White spaces and semicolons at the end of statements

White spaces are also ignored. Semicolons are used in some other programming languages to signal the end of a statement/instruction. In Python, they are not needed (they can be left out).

```python
#white space is ignored
print("without space between the added numbers: ",2+5)
print(    "with space between the added numbers: ",2  +   5)
```

```python
#But white spaces insided quotes (as part of strings) are NOT ignored
print("without space between the added numbers: ",2+5)
print("with    space between the added numbers: ",2  +   5)
```

```python
#semicolons at the end of a line make no difference
print("without space between the added numbers: ",2+5);
print(    "with space between the added numbers: ",2  +   5);
```

#### 6.2 Numpy package and working with arrays

Numpy is a numerical library for python that makes it easy to create and work with arrays. And we use arrays to be able to work with a sequence of values (e.g. a column or row of values in a table). Arrays make it possible to do calculations on a large number of values all at once. 

Arrays are created by putting a sequence of numbers separated by commas in square (not rounded!) brackets as we saw earlier. 

We can access values in an array by using `indexes`. Unlike in algebra, the first position in an array has an index of 0 not 1. See example below.


##### 6.2a Import Numpy and create an array 


```python
#first import the Numpy library and give an alias (shorter name of np)
import numpy as np
```

Jack bought 3 kilograms of bananas, 2 kilograms of oranges and 5 kilograms of lollies. The prices of the three items are 4, 6 and 9 dollars a kilogram, respectively. 

How much money did Jack spend on each item and in total? Solve the problem by first creating an array to store the quantities purchased and another array to store the corresponding prices. Find the answer by operating on the arrays you created.

```python
#create an array of purchased quantities
quantity = np.array([3, 4, 5])

#first item has an index of 0 not 1
print("Jack bought {} kgs of bananas, {} kgs of oranges and {} kgs of lollies.".format(quantity[0],quantity[1],quantity[2]))
```

```python
#create an array of corresponding prices
price = np.array( [4, 6, 9] )

#create an array of expenditures 
expenditure = quantity * price

#Jacks itemized expenditure is as follows
print("Jack spent {} dollars on bananas, {} on oranges and {} on lollies.".format(expenditure[0],expenditure[1],expenditure[2]))

#Then print total expenditure (use the sum() function in Numby to add the expenditures)
print("Jack spent {} dollars in total.".format( np.sum(expenditure) ) )
```

##### 6.2c Creating 2-D arrays and transposing 2-D arrays

A table of values is an array of arrays. For example, our data on purchased quantities and prices for Jack are two columns of data. We can create a 2-D array that includes quantities and prices as follows.

```python
#create 2-D array (as an array of arrays)
qp = np.array( [ [3,4,5], [4,6,9] ] )

#display: quantities are in first row and prices in second (commodities are in columns)
print(qp)
```

What are the dimensions of the table: 2(rows) x 3(columns) or 3x2?

```python
#check shape dimensions of the array
print(qp.shape)
```

Print only the data on lollies (quantity and price).

```python
#Select all (both) rows using the colon (:) value as row index and 
# the third (or lollies) column (which has an index of 2)

print( qp[ : , 2] )
```

However, in some cases we want to transpose the 2-D array. In the example above, for instance, a user might find it more natural to have the commodities in the rows and the values (quantities and prices) in the columns. In that case, we can use the *transpose()* function as shown below. 

```python
#transpose the array
qp2 = qp.transpose()
print(qp2)
```
