Variables store values. Python automatically determines the type.

## 3. Basic Arithmetic

Python can perform all standard mathematical operations:

```python
# Basic operations
a = 10
b = 3

print(f"Addition: {a} + {b} = {a + b}")
print(f"Subtraction: {a} - {b} = {a - b}")
print(f"Multiplication: {a} √ó {b} = {a * b}")
print(f"Division: {a} √∑ {b} = {a / b}")
print(f"Integer Division: {a} // {b} = {a // b}")
print(f"Modulus (remainder): {a} % {b} = {a % b}")
print(f"Power: {a}^{b} = {a ** b}")
```

## 4. Scientific Calculations

Let's calculate some real scientific quantities:

```python
# Kinetic Energy: E = 0.5 * m * v^2
mass = 2.5  # kg
velocity = 4.0  # m/s

kinetic_energy = 0.5 * mass * velocity**2
print(f"Kinetic Energy: {kinetic_energy} Joules")

# Population growth: N = N0 * (1 + r)^t
initial_population = 1000
growth_rate = 0.05  # 5% per year
years = 10

final_population = initial_population * (1 + growth_rate)**years
print(f"Population after {years} years: {final_population:.0f}")
```

## 5. Using the math Library

Python's `math` library provides additional functions:

```python
import math

# Square root
print(f"‚àö16 = {math.sqrt(16)}")

# Trigonometry (angles in radians)
angle_degrees = 45
angle_radians = math.radians(angle_degrees)
print(f"sin(45¬∞) = {math.sin(angle_radians):.4f}")
print(f"cos(45¬∞) = {math.cos(angle_radians):.4f}")

# Logarithms
print(f"log‚ÇÅ‚ÇÄ(100) = {math.log10(100)}")
print(f"ln(e) = {math.log(math.e)}")

# Constants
print(f"œÄ = {math.pi}")
print(f"e = {math.e}")
```

## 6. Lists: Storing Multiple Values

Lists hold collections of data - perfect for experimental results!

```python
# A list of temperature measurements
temperatures = [22.1, 23.4, 21.8, 24.2, 22.9, 23.1]

print(f"All temperatures: {temperatures}")
print(f"First measurement: {temperatures[0]}¬∞C")  # Indexing starts at 0!
print(f"Last measurement: {temperatures[-1]}¬∞C")
print(f"Number of measurements: {len(temperatures)}")

# Basic statistics
average = sum(temperatures) / len(temperatures)
print(f"Average temperature: {average:.2f}¬∞C")
```
